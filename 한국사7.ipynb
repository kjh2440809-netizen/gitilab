- -> 사화 = 훈구 vs 사림
- -> 붕당 = 사림 vs 사림
### 사림의 성장과 사화
- 성종 = 훈구파 견제 -> 사림 증용(김종직을 3사 언관직으로 데려옴)
> **언관 = 3사 - 비리 저지른 놈들 잡을 수 있는 말하는 관직**
- 사화의 종류 = 무오, 갑자, 기묘, 을사
## 연산군
- 무오사화(조의제문)
> 김종직이 세조(수양대군)을 비판하며 쓴 글, 세조는 연산군의 증조 할아버지임
> 연산군은 화나서 김종직의 제자들 능지처참, 김종직 부관참시(무덤에서 시체를 꺼내 목 자름)
- 갑자사화(폐비 윤씨 문제)
> 폐비윤씨는 연산군의 친 어머니임 질투가 강하고 문제가 있어 쫓겨남
> 그럼에도 반성하지 않아서 사약을 마시고 죽음
> 연산군은 어머니의 죽음에 대해 모르고 자랐음
> 사실을 알게 된 연산군은 화나서 갑자사화를 일으킴
## 중종
- 기묘사화
> 중종은 연산군(형)을 내쫓고 어린나이에 왕에 오름, 왕이 힘이 약해서 조광조를 등용
> 현량과 : 지방관리의 추천을 받는 과거시험
> 향약 보급
> 소격서 폐지(조광조는 사림이라 중국을 좋아함)
> 위훈 삭제 : 중종 반정의 공신들을 70% 짜름
> 조광조를 등용한지 4년째 힘이 너무 강해져 왕 따위 필요없다는 소문이 돌게됨
> 중종은 위기감을 느끼고(형처럼 될수도 있겠다) 조광조를 제거
- 을사사화
> 중종은 아내를 세번 맞이함
- 첫번째 부인
> 장인어른이 연산군의 처남이라 짤림
- 두번째 부인(대윤집안)
> 윤씨 집안에서 새로운 부인을 맞이함.
> 아들이 왕이 되는데 아들을 낳는 과정에서 죽음
- 세번째 부인(소윤집안)
> 같은 윤씨 집안에서 시집와서 인종을 키움
> 먼 친척인 대윤집안에서 소윤집안 사람들을 다 죽일거라고 중종에게 자길 죽여달라고 함
> 중종은 명종에게 왕위세습을 약속하고 8개월후 생을 마감함
> 명종이 왕이 된 후, 소윤집안은 대윤집안을 작살냄
### 사림의 기반
- 항약 : 상부상조 + 풍속유지 + 서로 감시,지도(조선시대 마을의 자치 규칙)
- 서원 : 지방 교육담당, 옛 선인에게 제사, 주변 선비들 대접 -> 향후에 붕당의 근거지로 변질
> 항약과 서원이 지방에서 사림의 근거지가 되어줬기 때문에 4번의 사화에도 버티고 주도권 잡음
