### 태종과 세조가 닮고, 세종과 성종이 닮음
-  | | 훈구파(관학파=국립학교 다님) | 사림파(사학파=사립학교 다님) |
|---|---|---|
|연원|혁명파(건국 공신, 세조 공신)|온건파|
|성격|자주적,주체적|존화주의적(성리학을 공부하다보니 중국을 좋아함)|
|정치 체제|중앙 집권 강화|왕권 건제, 향촌 자치|
|학문 경향|단구조선|기자조선(기자라는 중국인이 조선으로 건너옴 찬양함)|
|목표|부국 강병 -> 과학기술과 다른 사상 수용|왕도 정치 -> 성리학만 배움|

|시기|15세기|16세기 이후|
- -> 사화 = 훈구 vs 사림
- -> 붕당 = 사림 vs 사림
### 사림의 성장과 사화
- 성종 = 훈구파 견제 -> 사림 증용(김종직을 3사 언관직으로 데려옴)
> **언관 = 3사 - 비리 저지른 놈들 잡을 수 있는 말하는 관직**
- 사화의 종류 = 무오, 갑자, 기묘, 을사
## 연산군
- 무오사화(조의제문)
> 김종직이 세조(수양대군)을 비판하며 쓴 글, 세조는 연산군의 증조 할아버지임
> 연산군은 화나서 김종직의 제자들 능지처참, 김종직 부관참시(무덤에서 시체를 꺼내 목 자름)
- 갑자사화(폐비 윤씨 문제)
> 폐비윤씨는 연산군의 친 어머니임 질투가 강하고 문제가 있어 쫓겨남
> 그럼에도 반성하지 않아서 사약을 마시고 죽음
> 연산군은 어머니의 죽음에 대해 모르고 자랐음
> 사실을 알게 된 연산군은 화나서 갑자사화를 일으킴
## 중종
- 기묘사화
> 중종은 연산군(형)을 내쫓고 어린나이에 왕에 오름, 왕이 힘이 약해서 조광조를 등용
> 현량과 : 지방관리의 추천을 받는 과거시험
> 향약 보급
> 소격서 폐지(조광조는 사림이라 중국을 좋아함)
> 위훈 삭제 : 중종 반정의 공신들을 70% 짜름
> 조광조를 등용한지 4년째 힘이 너무 강해져 왕 따위 필요없다는 소문이 돌게됨
> 중종은 위기감을 느끼고(형처럼 될수도 있겠다) 조광조를 제거
- 을사사화
> 중종은 아내를 세번 맞이함
- 첫번째 부인
> 장인어른이 연산군의 처남이라 짤림
- 두번째 부인(대윤집안)
> 윤씨 집안에서 새로운 부인을 맞이함.
> 아들이 왕이 되는데 아들을 낳는 과정에서 죽음
- 세번째 부인(소윤집안)
> 같은 윤씨 집안에서 시집와서 인종을 키움
> 먼 친척인 대윤집안에서 소윤집안 사람들을 다 죽일거라고 중종에게 자길 죽여달라고 함
> 중종은 명종에게 왕위세습을 약속하고 8개월후 생을 마감함
> 명종이 왕이 된 후, 소윤집안은 대윤집안을 작살냄
### 사림의 기반
- 항약 : 상부상조 + 풍속유지 + 서로 감시,지도(조선시대 마을의 자치 규칙)
- 서원 : 지방 교육담당, 옛 선인에게 제사, 주변 선비들 대접 -> 향후에 붕당의 근거지로 변질
> 항약과 서원이 지방에서 사림의 근거지가 되어줬기 때문에 4번의 사화에도 버티고 주도권 잡음

> 효종 부인이 죽었을때 남인 1년 (둘다 1년이다)
> 1차 예송 : 효종이 죽었을때, 2차 예송 : 세자비가 죽었을때
- 인조는 큰아들을 두명 낳았는데 소현세자, 봉림대군임
- 하지만 소현세자는 죽고 봉림대군(효종)이 왕이 됨
- 그 아들이 현종인데, 현종과 세자비(현종 부인) 둘 다 있는 상황에서
- 현종과 세자비보다 어린 현종의 새엄마가 들어옴
## 숙종
- **환국 : 한개의 분당이 권력을 잡고있는데 그 분당을 싹 몰아내고 다른 분당에게 권력을 주는 것.**
- 경신 : 서인 vs 남인 = 서인
- 기사 : 서인 vs 남인 = 남인
- 갑술 : 서인 vs 남인 = 서인
> 기사환국 : 남인과 편먹은 장희빈이 숙종을 꼬셔서 남인이 주도권을 잡음
> 갑술환국 : 숙종이 젊은 동이를 좋아해서 꼬시다가 장희빈이 아니꼽게 봄
> 장희빈이 동이를 괴롭히는데 서인이 그걸 보고 동이와 주도권을 잡을 계획을 짬
> 동이가 숙종을 꼬셔서 서인이 주도권 잡는데 성공 -> 서인은 남인을 다 죽여버림
- 변질 : 일당 전제화 -> 왕권약화
- **일당 전제화 : 하나의 붕당이 모든 권력을 독점한 상태**
## 영조
- 왕권 강화를 위해 "탕평책"을 내놓음
- 완론 탕평 이라고도 함 = 완전한 탕평은 아니다.
- 탕평비를 세우고 붕당의 근거지인 서원을 정리함.
- 서원의 큰 스승들을 산림 이라고 하는데 산림을 부정함
> 중앙으로 보낸 제자들을 정치적으로 이용 못하게 함 = 이조전랑을 약화시킴
- 균역법 = 세금제도의 개혁(세금 줄어듬)
## 정조
- 노론이 주도권을 잡고 있는걸 인정하고 소론과 남인들으 중용 = 준론 탕평
- 규장각 설치
> 초계문신제 = 정조가 지방에 있는 하급 관원들에게 문재를 내서 규작각에 입학시킬 인재를 소집함
- 군사권을 장악하기 위해 장용영 이라는 군대를 만듬
- 수원 화성을 지음, 대전통평이라는 법전을 지음
