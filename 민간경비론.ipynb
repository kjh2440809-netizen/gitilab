# 민간경비의 의의
## 국가의 기능과 치안
### 국가의 기능
> 제1차적 기능
- 보호, 보장기능 : 국가의 안보기능(외침보호), 치안유지기능(질서유지)

> 제2차적 기능
- 복지기능 : 공동복지사업을 증진시킴

> 최근의 추세
- 1970년대 말 : 복지국가의 역할이 공공부문의 비대화, 비능률 및 재정위기로 위축되었음
- 1980년대 이후 :
- 개발도상국은 심각한 **경제적문제에 직면** -> 고비용의 공적 서비스분야의 미진함을 초래, **공적 서비스의 개혁이 정책과제로 부각됨** -> **공공부문에서 민간부문으로의 이양현상**

## 민간경비의 변천
### 대륙법계 국가의 경찰개념의 형성과 변천
#### **고대국가**
#### [경찰과 행정의 미분화]
> 경찰의 어원
- polis(도시국가)의 축소개념인 라틴어의 "politia"에서 유래
- politia : 시민의 생존과 복지를 보살피는 일체의 관리활동

> 경찰의 개념
- 헌법 또는 질서있는 공동사회를 의미
- 도시국가에 관한 일체의 **정치 = 헌법 = 경찰**
#### **중세시대(14~16세기)**
> 14세기 말
- 프랑스의 경찰(la police)개념 :
- 국가목적, 국가작용, 국가의 평온한 질서 있는 상태

> 15세기 말
- 독일의 경찰개념 :
- 질서유지를 포함한 국가행정 전반을 포괄하는 의미
- 공동체의 양호한 질서의 상태
- 프랑스의 경찰개념이 독일에 계승됨

> 16세기
- 독일의 제국경찰법상의 경찰개념 :
- 교회행정의 권한을 제외한 일체의 국가행정을 의미(중세는 '신'의 시대)
- 경찰개념이 세속적인 사회생활의 질서를 공권력에 의하여 유지하는 작용으로 축소
- 경찰권은 절대주의적 국가권력의 기초가 되었음(왕이 경찰권을 행사함)
#### **경찰 국가 시대(17~18세기)산업혁명**
#### [경찰과 행정의 분화]
> 등장배경
- 17세기 국가작용의 확대와 분화 :
- 국가활동이 **국가목적적 행정**(외교, 군사, 재정, 사법행정 등) 과, **사회목적적 행정**(소극적 질서유지와 적극적 복리증진을 포함)으로 분화
- 경찰의 개념 = 사회목적적 행정(적극적 복리증진 + 소극적 질서유지) = 내무행정 전반을 의미
- 국가목적적 행정은 경찰개념에서 **제외**
- 17세기에 **국가활동의 확대와 복잡화**로 나타난 국가작용의 분화현상이 **경찰국가 성립의 기반이 됨**

> 경찰의 개념
- 군사, 재정, 사법을 제외한/ **내무행정 전반**을 의미함

> 특징
- 이때 국가 행복은 돈 많이 버는것을 의미함 -> 소극적인 질서유지뿐 아니라 적극적 복리증진을 위해서도 경찰권 발동 가능 -> 국왕의 통치권이 내무행정 전반에 미침
- 경찰개념은 공동체의 양호한 질서상태가 아니라 **절대군주의 경찰권**을 의미함
- 군주의 명령은 **강제권** 특히 **형벌권**에 의거 집행되는 국가행정 전체를 의미
- 관료는 국왕의 절대적 권력에 복종하고 헌신하는 반면, **포괄적 권한**에 근거하여 **재판의 통제없이 국민의 권리관계에 간섭함**
- 경찰과 사법의 분리 : 사법이 국가의 특별한 작용으로 인정됨 -> 1648년 '베스트팔렌조약'에서 비롯됨
#### 19세기 법치 국가 시대
> 등장배경
- 독일에서 경찰국가 개념은 18세기에 칸트 등의 계몽철학이 등장하면서 극복됨
- 군주의 권력도 법에 구속을 받게 되는 법치주의 시대로 전환됨으로써 시민이 그 주체성을 회복함
- 자유주의적 자연법 사상과 권력분립주의가 대두됨

> 경찰의 개념
- 소극적인 질서유지 부터 그것을 위한 위해방지를 위한 작용 -> 적극적인 복리증진작용이 경찰개념에서 제외됨
#### 20세기 법치 국가 시대
> 등장배경
- 제2차 세계대전 후 보안경찰을 제외한 협의의 경찰사무가 다른 관청의 분장사무로 이관되는 비경찰화가 이루어져 오늘날과 같이 경찰의 임무가 보안경찰의 임무로 한정됨

> 경찰의 개념
- 공공의 안녕과 질서유지(강학상의 보안경찰)
#### 결어
- 경찰의 개념은 일체의 정치(=헌법) -> 국정전반 -> 내무행정 -> 위험방지 -> 협의의 행정경찰 -> 보안경찰 순으로 축소를 거듭

### 민간경비의 개념
#### 대륙법계
> 국가와 사회
- 대립관계 긍정

> 권력의 기초
- 통치권

> 경찰개념의 형성
- 발동범위를 제한해 나가는 과정(축소의 역사)

> 논의의 초점
- '경찰은 무엇인가'라는 문제로 경찰의 개념이 논의

> 경찰의 정의
- 사회공공의 안녕과 질서를 유지하기 위하여 일반통치권에 근거하여 국민에게 명령, 강제하는 내용

> 행정경찰과 사법경찰의 구분
- 구분(o) -> 사법경찰을 경찰의 고유업무로 보지 않음

> 경찰조직의 성격
- 중앙집권적 국가경찰

> 경찰의 임무
- 공공의 안녕과 질서의 유지에 중점(소극적 질서유지)

> 경찰의 수단
- 권력적 수단을 중시

> 경찰권의 의미
- 사회공공의 안녕과 질서를 유지하기 위하여 일반통치권에 근거하여 국민에게 명령, 강제하는 내용
#### 영미법계
> 국가와 사회
- 대립관계 부정

> 권력의 기초
- 주권자인 시민으로부터 위임받은 자치권한

> 경찰개념의 형성
- 경찰이 시민을 위해서 수행하는 기능을 중심으로 형성

> 경찰개념의 형성과정
- 발동범위를 **확대**해 나가는 과정 ->  적극적 복리증진을 경찰개념에 포함시킴

> 논의의 초점
- 경찰은 무엇을 하는가, 경찰활동이 무엇인가로 경찰의 개념이 논의

> 경찰의 정의
- 시민을 위하여 법을 집행하고 서비스를 수행하는 조직체

> 행정경찰과 사법경찰의 구분
- 구분(x) -> 행정경찰 = 사법경찰로 봄

> 경찰조직의 성격
- 지방분권적 자치체경찰(영국 52개, 미국 18,800개)

> 경찰의 임무
- 소극적 질서유지 + 적극적 공공복리 증진

> 경찰의 수단
- 비권력적 수단을 중시

> 경찰권의 의미
- 소극적 질서유지 + 공공복리 증진을 위해 개인의 자유, 재산을 규제(제한)하는 권한 -> 우리나라의 경찰권보다 넓은 개념

### 민간경비의 개념정의
#### 협의
- 국민의 생명, 신체, 재산, 보호, 질서유지를 위한 범죄예방활동 = 방범활동
#### 광의
- 협의 + 방재 + 방화 등
#### 최협의, 협의, 광의
#### 최협의
- 경비업체가 계약된 고객의 생명과 신체 및 재산을 보호하는 활동
- 자체경비를 제외한 **계약경비만**을 민간경비로 보는것
- 실정법에 의거한 경비활동 = 형식적 의미의 개념
#### 협의
- 민간이 주체가 되는 모든 경비활동
- 개인 및 기업의 자체경비, 계약경비(공공기관 X)
#### 광의
- 공공기관에 의한 공경비활동을 제외한 모든 경비활동
- 최광의로 이해할때는 공공기관이 민간인을 직접 고용해서 경비업무를 수행하는 혼성경비 활동까지 포함
#### 실질적
- 속성, 성질, 내용을 기준으로 한 개념
- **공공의 안녕과 질서유지** 등 **공경비가 수행하는 경비활동과 본질적인 차이 없음**
- 공경비와 **주체**면에서만 차이가 있음
#### 형식적
- **실정법** 규정에 따라 허가받은 법인에 의하여 임무를 수행
- 공경비와 민간경비가 명확히 구별됨
- 실질적 의미의 민간경비활동과 불일치하는 면이 많음
#### 영미법계
- 실질적 의미의 민간경비로 이해함
- 국민의 생명, 신체, 재산보호, 사회질서 유지의 주체로 경찰과 함께 민간경비도 당연한 것으로 인식함
- 민간경비의 업무범위와 공경비가  거의 유사하나, **법집행 권한에 대해서만 차이가 있음**
#### 대륙법계
- 전통적으로 국가권력의 우월적 지위를 인정함
- 경찰 : 일반(국민)통치권에 근거하여 국가행정의 수행이나 국가체제의 유지 등을 그 고유의 임무로 함
- 민간경비 : 경찰의 지도, 감독 하에 관련법규에 한정된 소극적 역할을 맡았고 사전적 예방적 기능만을 제한적으로 담당함
#### 미국의 민간경비의 일반적 역할
- 시설방범경비업무
- 질서유지업무
- 신변보호업무
- 화재예방 및 통제업무
- 비상 및 재난 통제업무
- 호송경비업무
- 정보조사 및 탐정업무 등
#### 한국의 민간경비의 역할 및 엄무
- 시설경비업무
- 호송경비업무
- 신변보호업무
- 기계경비업무
- 특수경비업무
- 혼잡, 교통유도경비업무
#### 민간경비원의 3가지 법적 지위(A. J. Bilek)
> 일반시민과 같은 민간경비원
- 한국에서 대부분의 민간기업체 민간경비원이 여기에 속함

> 특별한 권한을 가진 민간경비원
- 정부기관에 의해 특별한 경찰업무를 위임받은 민간경비원
- 제한된 근무지역 내에서 경찰권을 일부 사용할 수 있는 경비원
- 한국의 청원경찰이 여기에 속함

> 경찰관 신분을 가진 민간경비원
- 경찰관 신분으로 민간경비 분야에 부업으로 근무를 하는 등 두가지 신분을 가지고 일하느 민간경비원
- 한국은 아직 경찰공무원의 부업 또는 영리행위를 인정하지 않음


### 공경비의 개념
- 공익을 위하여 국민의 생명, 신체, 재산 보호, 공공의 질서유지, 범인의 체포, 수사의 권한, 범죄예방, 교통 통제와 같은 제반활동
#### 4차 산업혁명
> 노동시장의 양극화 심화
- 고급지식 노동자와 재산적 상위층은 부의 축적기회가 많아지지만 노동자들은 빈곤화가 심해질 것으로 에상하고 있다

> AI
많은 부분에서 로봇이 대체할 것으로 보고 있다

> 주요변화
- 초지능화 : 모든 사물이 점점 스마트해지는 것, 실제 현실의 사물제어가 가능한 수준이 되는 것
- 초연결화 : 세상의 모든 것이 연결되는 것
- 초감성화 : 미래는 감성을 기반으로 하는 창의성이 요구된다는 것을 ㅁ라한다
- 초융합화 : 전 분야에서 융합현상이 일어나고 있음을 의미, 학문 영역뿐만  아니라 산업 영역 전 분야에 걸쳐 나타나고 있는 특성이라고 볼 수 있다.
#### 민간경비와 다른 산업분야의 결합현상
- **조선, 자동차 등** 기타 산업과 결합되어 새로운 서비느사 제품의 안정성과 부가가치를 창출함
#### 융합보안 개념의 부각
- **물리적 보안 요소와 정보보안 요소**를 통합한 융합보안 개념이 부각되고 있음
- 국내 외에서 융합보안 개념을 새롭게 인식하고 있음
- 물리적 보안요소 + 정보보안 요소(기술적 보안요소 + 관리적 보안요소)를 상호 연계하여 통합적 접근방법이라 할 수 있음
- 개인, 기업, 정부단체 등의 데이터를 통합해 총체적으로 대응할 수 있다
#### 민간경비의 성장요인
- 정부의 형사정책적 대응 : 재범률이 비교적 높음
- 사회의 도시화, 과학화 발달에 의한 범죄의 복잡성
- 경찰력 증가와 범죄율 감소의 연관성을 발견하기 어렵고, 범죄예방과 통제에 유익한 영향을 주지 못하는 것으로 나타남 -> 경찰력 증가가 범죄율의 감소로 이루어지지 않는다는 것
- 경찰력에만 의존하던 의식에 변화가 나타나 민간경비에 대한 의존도가 높아지고 있음(->자경주의)
#### 브란팅햄(P. J. Brantingham)과 파우스트(F. L. Faust)의 3가지 범죄예방유형
- 1976년 브란팅햄과 파우스트의 3가지 범죄예방유형
> 제1차적 범죄예방모델
- 일반대중 대상
- 범죄를 저지르고자 하는 기회 자체를 제거하려고 함
- 성악설의 관점에서 바라보는 일반인 모두를 지칭

> 제2차적 범죄예방모델
- 우범자나 우범집단 대상
- 범죄가 발생 전에 예방하고자 하는 것
- 우범자란 견물생심 수준을 넘어선 범죄적 행동을 할 가능성이 높은 사람임
- 예방활동으로 경찰이 주최하는 스포츠행사, 교정기관 체험학습

> 제3차적 범죄예방모델
- 범죄자 대상
- 더이상 범죄를 저지르지 않도록 하기 위함
- 예방활동으로 형사사법기관의 활동에 의해 이루어짐

> 비용면으로 1차적 예방이 적고 3차적 예방이 가장 많은 비용이 듬

- 1986년 독일의 쿠베는 이 구조모델을 주사위 모델로 응용하여 이해하기 쉽도록 소개
#### 환경설계를 통한 범죄예방전략(CPTED)
- 환경적인 요소가 인간의 행동 및 심리적 성향을 자극한다는 것
- 범죄원인을 개인적 요인보다는 환경적 요인에서 찾으며 모든 인간은 잠재적 범죄욕망을 가진다고 보았음

> **CPTED**

- 뉴만의 방어곤간에 체계적으로 설명되고 있음
- 환경적 기회를 제거하는 직접적 통제활동
- 기본전략으로 자연적인 접근통제, 자연적인 감시, 영역성의 강화

- 기계적인 통제 : 장벽, 차단장치, 잠금장치 등

> 감시
- 조직적인 감시 : 경비원 배치, 순찰활동
- 기계적인 감시 : 경비조명, 감시카메라
- 자연적인 감시 : 창문(일종의 거리의 눈 개념)

> 최근에는 기계적 전략보다는 1차적으로 자연적 접근통제와 감시방법을 고려하고 2차적으로 조직적, 기계적 접근통제와 감시방법을 적용

### Maslow의 5단계 욕구이론
- 생리적욕구 -> 안전의 욕구 -> 사회적욕구 -> 존경의 욕구 -> 자아실현욕구 이며, 우선순위의 계층을 이룸
- 가장 낮은 (생리적 욕구, 안전의 욕구)는 사람에게 가장 필요한 것임
- 정상적인 사람의 궁극적인 목표는 자기실현 욕구
- 가장 강한 욕구는 생리적 욕구

> 생리적 욕구
- 의,식,주 및 건강에 관한 욕구(성적욕구)
- 충족요건 : 적정보수, 휴양제도

> 안전욕구
- 현재 및 장래의 신분이나 생활에 대한 불만을 해소
- 충족요건 : 신분보장, 연금제도
- 민간경비가 여기에 해당함

> 사회적 욕구
- 동료, 상사, 조직전체에 대한 **친근감, 귀속감**을 충족
- 충족요건 : 인간관계 개선, 고충처리, 상당

> 존경욕구
- 타인의 인정, 존중, 신망을 받으려는 욕구
- 충족요건 : 참여확대, 권한의 위임, 제안제도, 포상제도

> 자기실현 욕구
- 장래의 자기완성의 욕구 및 성취감 충족
- 충족요건 : 승진, 공무원단체 활용

#### Alderfar의 ERG이론
- 매슬로의 5단계 욕구를 3단계로 통합, 조정한것

> 존재욕구
- 생리적욕구 + 안전욕구
- 육체적인 생존을 유지하기 위한 모든형태의 생리적 욕구

> 관계욕구
- 애정,소속의 욕구 + 존경의 욕구
- 중요한 사람들과의 대인관계를 좋게 유지하고픈 욕구

> 성장욕구
- 존경의 욕구 + 자아실현의 욕구
- 자신감 + 자기발전을 추구하는 욕구





#### Herzberg의 2요인론
- 불만을 일으키는 요인과 만족을 주는 요인은 서로 다르다는 욕구충족 이원론을 제시

> 위생요인(불만요인)
- 불만요인이 충족되어도 적극적으로 만족감을 느끼게 하여 근무의욕을 향상시키지 못하므로 결과적으로 **단기간의 효과**를 가짐
- 직무의 '외'재적 성격

> 동기부여 요인(만족요인)
- 일단 충족되면 만족에 적극적으로 영향을 줄 수 있고 **일에 대한 적극적인 태도**를 유도할 수 있다
- 위생요인을(불만을) 해결해도 동기요인(만족)을 작동시키지 못한다[불만은 해결해도 그냥 평상시 상태일 뿐이고 만족하진 않는다]

> 위생요인과 동기부여요인은 상호 독립되어 있다

- 타워팰리스는 화재, 지진, 강풍, 범죄 등의 다양한 복합위험성을 종합적으로 관리할 수 있는 종합경비 시스템을 구축하고 있음

#### 경제적 관점
> 경제환원이론

- 거시적 차원에서 범죄의 증가를 경제침체로 인한 실업의 증가에서 찾으려고 하는 것

> 수익자부담이론
- 경찰이 갖는 근본적 성격과 역할 및 기능에 대한 통념적 인식에 의문을 제기함
- 경찰의 역할이 개개인은 안전과 사유재산을 보호하는 것이라는 관념을 거부함
- 행정수비범위론과 민간경비차원에서의 수익자부담과도 밀접한 관련성이 있다고 봄
- 경찰의 공권력 작용은 거시적 측면에서 질서유지나 체제수호 같은 역할과 기능으로 한정함
- 개개인이나 집단등의 안전과 보호는 해당 개인이나 조직이 담당하여야 한다는 인식에 민간경비의 수요증대는 불가피함
- 수익자부담이론의 성립조건 : 국민 소득의 증가, 범죄의 증가, 민간경비 서비스의 유용성에 대한 인식변화
#### 사회적 관점
> 공동화이론
- 민간경비가 경찰 본연의 기능 또는 역할을 보완하거나 대체한다는 이론
- 범죄증가에 경찰력이 증가하여야 하나 현실적으로 힘들고 그 결과 생겨나는 갭을 매우기 위해 민간경비가 발전한다는 이론

> 이익집단이론
- 그냥 내버려두면 보호받지 못한 채로 방치될 재산을 민간경비가 보호한다
#### 행정적 관점
> 민영화이론
- 미국의 민영화에 대한 논의는 1970년대 이후 **신자유주의** 확대에 따른 정치변화의 바람을 크게 탐 -> 민간경비의 성장으로 이어짐
- 근대국가의 성립 이후 국가의 본질은 물리적 강제력을 독점적으로 사용하는 것임 -> 형사사법과 군사 등의 영역에 사적인 개입을 허용하지 않음
- 신자유주의 경제학자들을 중심으로 관료제정부의 축소와 시장경제로의 환원의 요구가 많아졌고 결과적으로 치안기능이 민간화의 대상이 됨
- 정부의 비용절감을 위하여 공경비의 역할을 줄이는 대신 민간경비의 역할이 확대된다

> 공동생산이론
- 치안서비스 생산과정에서 경찰과 같은 공공부문의 역할 수행과 민간부분의 공동참여를 통해서 민간경비가 성장함
- 경찰역할의 한계점을 일부 극복하고 시민의 안전욕구를 증대시키기 위해 시민 자신이 주체가 되는 능동적 참여를 다각적으로 유도하는 이론

> 민영화의 한계와 치안서비스 주체로서의 시민 참여
- 민영화는 일반시민의 참여를 제한하여 시민정신의 함양을 중시하는 **미국의 정치적 이념과 충돌**하는 결과를 야기함
- 정부기관과 시민이 서비스 공급의 책임과 권한을 공유함으로써 공동생산방식에 대한 연구와 노력이 시작됨

> 참여주체의 다양화
- 치안수요가 증가되어 경찰과 민간경비의 상호협력을 형성해 나가고 있다
- 민간경비의 지속적인 급성장을 중요하게 고려하여 자체적인 중요성이 전반적으로 수용되기 시작하였음
- 민간경비를 경찰활동의 보조가 아니라 치안서비스의 주체로서 인식하고자 함 -> 치안서비스 생산주체의 다원화
- 시민과 민간경비가 독립된 생산주체로서 참여하게 됨

> 공동생산에 의한 민간경비의 역할
- 경찰 : 거시적으로 공공의 안녕과 질서유지
- 시민참여 : 미시적인 개인 및 거주지역의 안전확보
- 민간경비 : 영리추구는 본질적 목표, 치안서비스 생산은 목표달성을 위한 수단

> 시민들의 공도생산 참여 유형
- [개인]
- 소극적 : 방범장비의 휴대, 이중자물쇠 등
- 적극적 : 범죄신고, 범죄정보 보상프로그램
- [집단]
- 소극적 : 아파트 공동경비
- 적극적 : 자율방범대, 이웃감시활동

> 지역사회 경찰활동 이론
- 전통적 경찰활동의 실패 -> 지역사회 경찰활동이라는 추세를 이끌었음
- 지역사회와의 신뢰관계를 형성하여 적극적으로 지역사외의 문제를 파악하여 지역주민과 협력하여 해력하려는 것
- 사후 대응보다는 사전 범죄예방을 선호하고 경찰활동에 시님의 적극적 참여를 유도함으로서 경찰의 한계를 극복

> 정부 실패론
- 치안문제에 대한 개인의 직접 통제 :
- 민간부문이 정부실패 내지 경찰실패의 상황에서 고유권리에 대해 관여, 회수하려는 건 오히려 자연스럽다













### 민간경비의 공공성과 경찰의 관계
#### 민간경비의 공공성
> 공공서비스의 개념
- 공공기관의 재화나 서비스로 인간의 문제를 해결하고 바람직한 환경을 만들려는 활동

> 치안서비스의 개념
- 공공의 안녕과 질서를 유지하면서 범죄와 무질서 각종 재해 등의 위험으로부터 보호

> 치안서비스 제공의 주체
- [민간부문]
- 경찰과 마찬가치로 범죄예방에 국한되지 않고, 각종 손실예방과 관련하여 수많은 역할을 수행함
- 민영교도소도 민간경비의 한 형태로 볼 수 있음








#### 치안서비스의 성격
- 공공성 정도가 강할수록 민간부문보다 정부에서 제공하는 것이 바람직함

> 공공성 정도에 따른 재화와 서비스의 구분
- 순수공공재
- 비순수공공재
> 공공부문과 민간부문의 중간영역에서 제공되는 재화와 서비스
- 민간재
##### 머스그레이브가 제시한 순수공공재의 3가지 기준
- 비경합성
- 비배제성
- 비거부성

> 비경합성 공동소비
- 소비하는 사람 수가 많아도 양에는 전혀 변함이 없다
- 추가이용자의 추가비용이 발생하지 않는다
- [민간재화]와 반대되는 성질

> 비배제성
- 특정 집단 또는 개인을 서비스 제공대상에서 제외시켜 혜택을 받지 못하도록 할수 없음
- 이용 또는 접근에 제한할 수 없다

> 비거부성
- 일단 공급된다면 모든 시민들은 자신들의 의지와는 상관없이 서비스를 이용하게 됨
##### 경찰서비스의 순수공공재성의 한계
- 현실적으로 순수 공공재에 해당되는 것은 희소하고, 대부분은 비순수공공재 또는 준공공재(=순수공공재와 민간재의 중간영역)에 해당함
- 공공재는 소비에 있어서 비경쟁적 성격을 가짐
- 비순수공공재는 공급량이 일정할때 한 사람의 추가사용은 타인의 사용을 배제하지는 않으나 효용을 감소시킴 이런부분을 정체현상 이라고 함
- 경찰서비스에 대한 지리적 관점에 대한 논의가 중요한 이유는 지리적 관점이 공공서비스가 갖는 순수공공재이론을 가장 침해하는 요인이 되기 때문임
- 티보트와 타이츠는 지리적 요인을 행정구역의 편성, 거리감퇴효과, 외부효과성 세가지 차원에서 설명함

> 행정구역의 편성
- 대한민국의 경찰은 국가경찰체제를 근간으로 하나, 경찰활동은 경찰청(전국적) -> 지방경찰청(시,도) -> 경찰서 -> 지구대(파출소)의 단위 중심으로 이루어짐
- 경찰서비스는 비순수공공재성을 가지고있음, 관할지역이 작아질수록 순수공공재성을 유지하기 어려움

> 거리감퇴효과
- 행정구역안에서 공공서비스들은 무료로 이용할 수 있으나 현실적으로는 관할 지역주민들이 비용, 시간, 노력 등을 부담해야 함
- **치안서비스의 양과 거리의 관계 = 반비례관계**
- **거리와 서비스 공급비용의 관계 = 비례관계**
- **지안서비스의 질과 거리의 관계 = 반비례관계**

#### 경찰서비스에 대한 민간경비의 참여 가능성
- 경찰서비스가 특정지역을 제외한 곳에서는 잘 이루어지지 않기 때문에 **민간부문**의 참여가 논의 됨, 지역사회 경찰활동 또한 재해석이 필요함
- 민간경비가 공공서비스로서의 본질적 성격은 영리(돈) 목적이고, 생산주체가 치안환경 개선의 상품으로 어느정도 생산, 제공하기 때문임











### 민간경비와 공경비의 활동 영역
#### 뉴만의 방어공간 4단계 계층화
- 공적 영역 : 아파트 단지 **밖의 길거리**
- 준공적 영역 : 아파트 단지 **내의 정원이나 놀이공원, 주차장** 등
- 준사적 영역 : **복도 및 엘리베이터**(거주자들이 공통으로 사용하는 공간)
- 사적 영역 : **자기 집**(거주자가 자유롭게 생활하는 공간)
> 타워팰리스등 고급 아파트단지는 위의 영역설정이 명확하게 적용될수 없음 준공적 영역까지 순찰활동이 이루어짐
#### 베타성의 정도
- 배타성은 자신이 생활하는 공간에 어느정도 영향력을 행사할 수 있가가 하는 문제임
- 사적영역은 적용됨, 준사적 영역부터는 베타적인 권한이 제한됨
#### 민간경비와 공경비의 활동 정도
- 공적영역 : 공경비
- 준공적 영역 : 민간경비 < 공경비
- 준사적 영역 : 민간경비 > 공경비
- 사적 영역 : 민간경비















### 민간경비와 경찰의 역할관계 및 관계구조
#### 민간경비 - 경비업법 제2조
- 시설경비업무
- 호송경비업무
- 신변보호업무
- 기계경비업무
- 특수경비업무
- 혼잡, 교통유도경비업무
#### 청원경찰 - 청원경찰법 제3조, 동법 시행규칙 제21조 제2항
- 청원경찰은 청원주와 그 구역을 관할하는 경찰서장의 감독을 받아 경비구역만의 경비를 목적으로 경찰관의 직무를 수행함, 단 수사활동을 수행해서는 아니 된다
#### 경찰
- 국민의 생명, 신체 및 재산의 보호
- 범죄의 예방, 진압 및 수사
- 범죄피해자 보호
- 경비, 주요 인사 경호 및 대간첩, 대테러 작전 수행
- 공공안녕에 대한 위험의 예방과 대응을 위한 정보의 수집, 작성 및 배포
- 교통단속과 교통 위해의 방지
- 공공의 안녕과 질서 유지(가장 상위의 개념)
#### 구 경찰방문 및 방범진단 규칙
- 경찰방문이란 관할구역 내의 각 가정, 상가 및 기타시설을 주민의 협조를 받아 방범진단을 하는 예방경찰을 말한다
- 방범진단이란 범죄예방 및 안전사고방지를 위해 미비점을 보완하도록 지도하거나 경찰력 운용상의 문제점을 보완하는 활동
- 방범심방 : 경찰이 각 가정, 기업체, 기타 시설을 방문하여 범죄예방, 선도, 안전사고방지 등에 대해 지도, 계몽하는 활동
#### 자율방범대
- 봉사활동을 하는 단체로 지역주민들이 자발적으로 경찰서장에게 신고한 단체
#### 민간에 의한 방범활동
- 자율방범대에 의한 방범활동
- 시민단체에 의한 방범활동
- 언론매체에 의한 방범활동
#### 한국의 민간경비와 경찰의 제 관계
- 민간경비의 내용에 있어서 자체경비보다는 계약경비를 중심을 ㅗ비교함
- 자체경비를 포함시킬 경우 개인과 기업도 해당되고, 재원도 경비업법상 자본금과 경비료라기보다 개인 및 기업이 가지고 있는 자본금에 의함

> 공통점
- 범죄예방
- 질서유지
- 손실감소, 범죄감소, 위험방지
- 동일한 안전기법 활용

> 차이점
- 주체 : 공경비 = 국가,정부 , 민간경비 = 민간영리법인
- 대상 : [공경비 = 불특정 일반국민, 시민지향적 서비스] , [민간경비 = 특정 의뢰인 및 단체, 고객지향적 서비스]

> 경쟁적 관계구조
- 스코트와 맥퍼슨은 서로 경쟁관계에 있다는 입장임, 치안서비스에 있어서 기본적으로 차이가 있는것은 아님(법적 권한 빼고)

>협력적 관계구조
- 이익집단론은 민간경비를 하나의 독립적인 행위자로 인식함
- 민간경비의 본질적인 존립목적 = 영리지향성 -> 치안서비스는 상품으로서 존재함
- 특별권한의 인정 :
- 영국 = 호송업무 권한 부여
- 미국 = 경찰과 유사한 권한을 부여하는 특별경찰 제도
- 한국 = 특수경비원들에 대해 총기사용 권한 인정












# 민간경비의 발전과정
## 민간경비의 역사적 발전 배경
### 고대
#### 원시시대(생존을 위한 경비)
- 민간경비가 공경비보다 먼저 발전됨
#### 함무라비시대(국가적 차원의 경비)
- 법집행 개념의 등장 : 최초로 명문화됨으로써 이때부터 공경비 개념이 출현함
- 국가차원의 공경비와 민간차원의 민간경비의 분리가 시작됨
#### 그리스시대(국가경찰개념 등장)
- 부족이나 씨족 차원의 경비개념에서 사회차원의 공공개념으로 발전함
- 스파르타에서는 법집행을 위한 치안책임자제도가 발달함 -> 최초의 국가경찰의 발달로 추정됨
#### 로마시대
- 지배자가 통치하는 군대조직에 의한 사회질서 유지함 -> 국가적 차원의 경비
- 최초의 비무장 수도경찰로 간주됨(기원전 27년 아우구스투스 황제의 자경단원)

> 고대사회 이후 개인의 생명, 신체, 재산 보호수단 변화과정
- 자기보호 -> 부락공동보호 -> 국가보호 -> 자기보호와 국가보호

> 경비 개념의 시대별 분류
- 민간경비시대 -> 공경비와 민간경비 개념의 **미분화**시대 -> 공경비시대 -> 경경비와 민간경비 **병행**시대

- 위와 같은 발달과정은 함무라비왕조와 로마역사에서도 나타남

> 현대
- 영,미법계 = 공경비와 민간경비가 동등하게 발전함
- 대륙법계 = 공경비 위주의 체제로 발전함














## 각국의 민간경비
### 영국의 민간경비
#### 원시경비형태(1066년 노르망디 정복 이전)
> 드루이드 종교회의
- 잉글랜드 웨일즈 지방의 법집행의 시발점
- 겔트족의 승려 또는 목사회의로 공.사문제를 결정함 -> 중범죄자는 사형, 경범죄는 행사 제외

> 민중의 평화제도
- 관습적으로 전해지던 공동지역방위체제 -> 집단안전체제
- 모든 사람은 자기 지역사회의 구성원의 과오에 대하여 공동으로 책임을 지며 범법자에 대하여는 복수할 수 있다

> 프랭크 플리지 제도 = 티싱(Tithing)제도
- 12세 이상의 부족원들로 구성된 다가족부대를 기본단위로 하는 지역공동방위개념 또는 시민공동책임제
- 국왕의 평화를 달성하기 위한 경찰의 원형
- Tithing(티싱)[10인 조합] : 10가족으로 구성 -> 그 중 한명을 10인 조합장으로 선출 -> 10인 조합장은 도적을 규환(Hue ang Cry) 범죄자를 체포, 형벌부과할 책임을 가짐 -> 한 지역의 법과 질서유지르 위해 집단적 연대책임을 부화하는 비공식적 사회통제제도로 변화함
- Hundred(100인 조합) : 10개의 티싱으로 구성됨
- 함성제도 (Hue ang Cry) :
- 연대책임, 범죄를 발견한 즉시 고함질러서 알림
#### 노르만 정복시대
> 장원제도
- 1066년 노르망정복 이후 성립된 봉건체제하에서 법집행을 담당한 기본단위

> 장원의 영주
- 왕의 지방대리인으로서의 기능을 담당함
- 백인장의 역할이 샤이어즈나 샤이어리브로 점차 전환되어 감(지휘관 -> 국왕의 신하) -> 다시 주장관인 쉐리프로 발전함

> 레지스 헨리시법의 공포(1116년) : 헨리1세
- 공경비 개념의 등장 = 경찰은 공복으로서 더이상 사립경찰 활동을 하지 않도록 함

> 국가소환제도
- 주민의 규환제도가 발전하여 제도적인 정착을 한 것
- 범법행위에 대해서는 국각가 법으로 다스린다는 내용
- 국가소환제도는 오늘날의 공경비의 개념 <-> 주민소환제도가 완전 민간경비의 개념
- 모든 범죄는 국왕의 치안의지에의 도전으로 국왕에게 37종에 달하는 재판권을 부여함
- 중죄와 경죄에 대한 법률적인 구별
- 법개념이 사법에서 공법으로 변천하는 과정임
#### 중세(1300~1500)
> 윈체스터법 = 제2차 웨스터민스터법(에드워드 1세)
- 티싱제도의 법규상 의무화 = 평화유지를 부락책임으로 하는 원칙을 법제화
- 오늘날 한국의 경찰관 직무집행법과 형사소송법의 내용을 함께 포함함
- 중세 영국 경찰조직의 기본법
- 치안판사의 지휘감독하에 치안관이 담당함
- **주야간 순찰(watch and ward)**
- watch는 야간순찰을 의미, ward는 주간순찰을 의미함
- 도시의 남자들이 실시
- 야간경비원은 일몰부터 일출시까지 출입문주변에 배치
- 범인의 호창(hue and cry)

> 무기소지규정
- 과거 1181년 헨리 2세의 무장조례 제정 : 왕에게 인정받은 특정 계층들만이 무기소지가 가능했고 이들은 왕에게 충성을 맹세해야 함
- 에드워드 1세 : 15~60세까지 모든 남성들이 집안에 무기를 보관하도록 의무화함

> 상인경찰
- 영국의 세계통상이 활발해지면서 상인들의 재산보호와 범죄수사를 위한 사설경찰로 시작하여 발전함
- 치안부재의 결과는 민간경비의 발달로 이어지고 이들은 주,야 파수 등으로 일함
- 상인들이 도난당한 물건을 회수하기 위해 사람을 고용한게 오늘날의 사설탐정으로 발전하게 됨

> 교구경찰
- 도시는 교구로 나뉘어져 있으며 범죄가 증가함에 따라 교구경찰을 조직함
- 전 교구민이 윤번제로 근무함 -> 후에 교구치안관으로 대치함
#### 근대 이후
> 헨리 필딩
- 바우가 경찰대 3개 부문 조직
- 도보경찰대 : 도심지 근무
- 기마경찰대 : 도심에서 15마일 떨어짐
- 절도체포대 :
- 도난당한 재산의 회수활동, 자원봉사자들로 구성됨 시민 스스로 단결해야한다는 개념을 창시
- 1785년 경 최초의 형사기동대로 활동함
- 1829년 수도경찰에 흡수됨

> 로버트 필
- 산업혁명과 과학기술의 발달 -> 치안수요의 증가 -> 경찰제도 개혁 필요(공경비제도 출현)
- 로버트 필경은 1829년 수도경찰청법을 의회에 제출 -> 수도경찰청 창설 -> 민간경비와 공경비가 병행,발달하는 계기가 됨
- 경비대, 바우가 경찰대, 자경대 등을 하나의 능률적인 유급경찰로 통합함
- 근대 수도경찰조직을 만든사람
- Peeler 또는 Bobbies라고 불리움
#### 현대적 의미의 방범활동
- 영국 치안 = 공공부문 + 민간부문
- 민간치안기구의 경비유형 : 자체경비,계약경비 형태 , 기계경비용 장비판매업
- 민간경비산업의 유형 : 인력경비, 기계경비, 조사업무
#### 영국 민간경비산업 위원회 및 의무적 자격증 제도
- 독립된 민간경비산업위원회를 두어 규제업무의 전문화와 품질향상 효과를 높밍으로써 자율적인 발전을 이끌고 있다
- 민간경비산업의 의무적 자격증 부여와 자체적인 기준에 따라 민간경비업자를 평가하는 자발적인 인증계약제도를 관리한다
- 영국의 민간경비산업위원회는 민간경비업법에 근거해 2003년 설립되었고 내무부에 의해 제정지원을 받고있다
- 자격증의 종류는 현금호송, 신변경호, 출입감시, 열쇠관리, 공공감시, 일반경비, 차량견인 및 도난 방지
- 인증계약자제도를 심사하고 등록할 수 있는 의무 및 권한을 가지고 있다










### 미국의 민간경비
#### 역사적 배경
> 연혁(변천사)
- 영국왕실의 권위주의적 통치방식을 배척하고 자치적인 지방분권주의적 통치방식을 선호함
- 식민지 시대의 미국은 영국으로부터 범죄예방체계가 상당 부분 계수됨
- 공공경찰조직 설립 순서 :
- 1838년 필라델피아 경찰국, 보스톤 경찰 탄생
- 1845년 뉴욕 -> 미국에서 최초의 현대적 의미의 경찰조직이 설립된 곳

> 특징
- 국가통치제도 : 자치제도를 추구함
- 시민 스스로가 자경사상에 입각함, 지역 실정에 맞게 각기 특색있는 제도로 발전함
#### 민간경비의 발전과정
> 건국 초 북부지역
- 상.공업의 발달, 인구의 급격한 증가 -> 급격한 질서붕괴, 범죄증가
- 공적 법집행조직 : 경찰조직 -> 범죄예방이나 순찰 등의 기능은 미흡함
- 민간경비조직 : 자경단과 유급경비원이 야간경비나 순찰 등의 임무를 수행함

> 건국 초 남부지역
- 남부 농장지대에서 노예제도가 발달함 -> 농장노동자(노예)의 탈출, 소요사태 등 통제의 필요성이 나타남
- 공,사경비조직 모두 노예나 농장노동자들의 통제 필요성에서 발달됨, 노예 진압, 탈출한 노예의 체포
- 도망노예송환법 : 도망간 노예를 체포하여 원래의 주로 돌려주도록 규정한 법률
- 1793년 북부 주는 이 법에 대해 거세게 반발했으며, 일부 북부 주는 개인의 자유에 관한 법을 제정해 도망노예가 배심원 재판을 받을 수 있도록 했음
- 1850년 제2차 도망노예송환법 : 남부의 요구로 제정되었음, 도망노예는  자신을 위해 증언할 수도 없을뿐 아니라 배심원 재판도 받지 못하게 되었음, 법의 집행을 거부하거나 노예를 놓친 연방보안관들에게는 높은 벌금이 내려졌으며 노예 탈주를 도운 사람들에게도 벌금이 부과되었으나 이 법은 너무 가혹하게 시행되어 남용을 가져옴 -> 따라서 노예제 폐지론자들의 수가 늘어나 북부의 많은 주에서는 개인의 자유에 관한 법이 새로 제정되었음

> 건국 초 서부지역
- 19세기 중엽 골드러쉬로 표현되는 서부개척시대. 동부의 주민들이 서부로 이주 시작함
- 서부지역 토착 인디언의 공격과 도적들로부터 위협, 동부에 치우쳐 있던 공권력은 서부에 미치지 못함
- 민간경비조직 : 서부개척시대 이주민의 금괴수송을 위한 자경조직 철도경비를 위한 경비회사 발달 -> 철도경찰
- 1851년 최초의 민간경비라고 할 수 있는 핑커톤 국가형사국을 설립하여 철도경비를 전문으로 하는 경비업무를 수행함
- 철도회사들은 각 회사 소속으로 철도경찰대를 창설할 수 있게 됨
- 철도 노선에서 경찰권을 행사할 수 있었음
- 미국의 정부기관들과 법집행기관들이 설치된 후 자리를 잡을 때까지 주민들의 생명과 안전을 책임지는 중요한 역할을 수행함
- **엘런 핑커톤**
- 미국 철도경찰의 발전에 커다란 기여를 한 사람, 미국 민간경비산업에서 가장 대표적인 인물
- 위폐사범 일당을 검거하는데 결정적인 공헌을 하여 부보안관으로 임명되었음 -> 1843년 시카고 최초의 사법경찰관으로 임명됨
- 1850년 사법경찰관직을 사임하고 미국 최초로 철도경찰 업무를 수행함 -> 시초의 현대적 의미의 민간경비
- 1851년 핑커톤 (경비회사, 국가형사국, 탐정사무소)를 설립
- 미국에서 유일하게 주간 철도수송의 안전을 도모하는 업무를 담당하는 경비회사가 됨
- 경찰의 치안능력이 현저히 부족한 상태에서 핑커톤의 용의자 관련 자료와 정보는 바로 공공질서와 치안의 원천이었음
- 정부 어느 기관에서도 범죄자 정보를 기록, 관리하는 곳이 없을때 핑커톤은 유형별로 범죄자 기록을 정리, 오늘날 프로파일링의 전형을 세움
- 경찰의 요청이 있을 때 제공함으로써 경찰과 민간경비업체의 바람직한 관계를 정립함
- 남북전쟁당시 핑커톤사의 탐정들이 조사한 결과 실제로 링컨 대통령 암살계획이 비밀리에 진행되고 있음을 밝혀냄 -> 핑커톤이 링컨과 동승하여 기차를 바꿔타서 링컨을 암살위험으로부터 구함
- 홀메스 = 에드윈 홈즈

> 19세기이후 20세기 전반

> 경찰관 신분의 민간경비원
- 일반경찰관이 비번일때 민간경비원으로 일하는 것

> 미국의 경찰관 = sworn police officer + peace police officer
- sworn police officer : 비번일때에도 총기휴대, 사용 가능함
- peace police officer : 비번일때에는 일반 시민의 지위로 돌아감

> 민.형사법상 지위
- 미국의 불법행위법은 민간경비원에게 특별한 권한을 부여하지 않으며 민간경비원의 행위에 대하여 어느 정도 제한을 규정함
- 민간경비원의 불법행위도 일반인의 불법행위와 동일한 법리에 의하여 민사책임을 부담하고, 계약법상 엄격히 규제함
- 민간경비의 법적 문제는 주의 입법이나 지방조례 등에 허가된 민간경비원에게 특정조건 하에서 권한을 부여하고 있음
- 형사법상 지위 :
- 실력행사의 경우 : 과도 또는 비합리적인 폭력행위 발생시 손해배상책임과 본래의 특전도 상실됨
- 심문과 질문 : 일반시민이 법적으로 억류되어 있는경우 질문에 대해 반드시 응답하여야함
- 수색 : 경찰관의 협조 하에 행동하지 않는 한 민간경비원의 행위는 일반시민과 동일함








### 일본의 민간경비
#### 초기(도쿠가와 막부시대 이전)
- 중세부터 지방성주들의 사적으로 경비 실시
- 유력성주가 지방영주들을 1개월에 10일씩 상경하여 숙직, 경위, 잡역의 봉사를 행하게 함
- 도쿠가와 막부시대 : **장병위**라는 경비업 전문 직업경비업자가 등장
- 헤이안 시대(794~1080)에 출현한 무사계급에서 그 뿌리를 찾을 수 있음. 사무라이 라고도 함
#### 후기(도쿠가와 막부시대 이후)
- 장병위 경비업무를 넓혀 자택경비, 물품 운반까지 실시됨
- 제2차 세계대전 전 : 청원순사제도
- 공공기관에 의한 사회치안이 잘 되어있었기 때문에 관심은 높지 않았음
- 일본의 민간경비 역사가 짧은 이유는 물과 안전은 그다지 중요하지 않다는 의식이 강했음
- 제2차 세계대전 전에는 일본 산업계가 야경, 숙위, 순시, 보안원 등의 자체경비를 실시함
- **청원순사제도** : 극히 일부지만 경비를 민간기업에서 부담하고 경찰력의 상주파견을 요청하는 제도
- 일본은 전쟁수행을 위해 군수공장을 가지고 있었고 경비를 강화할 필요가 있었음, 그러나 국가는 돈이 없었기 때문에 군수시설의 시설주에게 소요경비를 부담하게 하고 경비책임을 지우는 청원순사제도를 만들어 청원순사로 하여금 관련시설의 경비를 맡도록 함
- 위탁 및 계약방식에서 합리적인 경비시스템의 개발을 막아 순수 민간경비제도의 발전에 저해요인으로 작용됨
- 현재 **한국의 청원경찰제도와 유사함**
#### 현대적 의미의 민간경비
- 제2차 세계대전 이후 급격한 산업화와 범죄의 증가에 따른 민간경비의 필요성이 사회적으로 증가하기 시작함
- 특정계층, 특정대상에 실시되어 오던 민간경비는 현대적 민간경비 관련 업무를 전문으로 하는 민간경비가 탄생하게 되었음 -> 1962년 7월 발족한 일본 경비조장주식회사가 세콤의 전신임
- 스웨덴의 경비회사와 제휴하여 도입한 것
- 1964년 동경 올림픽 선수촌 경비를 계기로 민간경비업체는 그 활약과 역할이 대단한 것으로 평가됨
- 1970년 만국박람회가 일본 오사카에서 무사히 개최됨 -> 시설관리, 안전, 질서유지등에 민간경비가 투입됨 -> 일본 민간경비의 양적 질적 성장을 보여줌
- 1972년 경비업법 제정 민간경비원 약 10배 증가
- 1972년 경비업법 : 규제사항을 정하고, 경비원에 대한 특별한 지위는 일체 인정하지 않는다
- 경비원의 결격사유가 경비업을 영휘하려는 자의 공안위원회에 신청(=신고) 의무
- 1982년 개정됨 : 부적격자를 배제하기위해 경비업을 영위하려는자의 요건을 신고제에서 허가제로 강화시킴 -> 경비업을 영위하려는 자는 결격사유가 없이 도도부현(광역)공안위원회의 인정을 받아야 함, 기계경비가 활성화 됨
- 일본의 민간경비의 발달은 고도의 경제성장과 범죄발생률 증가와 함께함 -> 1980년대 한국,1988년 중국 진출
- 1990년대 전후 일본 민간경비의 동향 :
- 상주경비 -> 기계경비
- 과점화와 양극분화 : 세콤과 종합경비보장의 과점화가 매년 심화됨
- 기계경비의 경우 2대 그룹 본사만으로 과반수를 초과하는 시장점유율을 차지함 -> 활동분야가 넓고 수요가 증가하면 상당한 저비용화가 가능함

> 경비택시
- 사고, 응급환자의 발생 등 긴급사태가 발생했을때 택시가 출동하여 관계기관에 연락하거나 가까운 의료기관에 통보하는 제도, 본래업무에 지장이 없는 범위에서 경비업과 대행업을 겸업

### 독일의 민간경비
#### 민간경비업 출현과정(고대~19세기 말)
- 최초의 문헌상 기록은 1430년경 화재예방과 화재진압
- 1870년대 최초로 도시단위의 야경조직들이 국가단위로 합쳐짐
- 1892년 경찰비용법이 통과되어 비용문제가 경찰행정예산으로 처리 됨
- 공공의 안녕과 질서업무가 국가 사무화 되는등 중앙집권화 경향은 경비업 출현에 결정적 전제조건이 됨
#### 최초의 민간경비회사 출현 이후
- 1901/07/15 Jacob의 고향도시인 Hannover에서 독일 최초의 민간경비회사를 설립하였다
- 1904년에 설립된 회사는 입법을 통하여 경비업을 위한 영업법을 요구하는데 찬성하였다
- 통일직후 구 동독지역에서는 서독지역에서 상업중이던 회사들이 대거 진출하였다
- 경찰관, 군인, 국경경비대 등을 경비회사 직원으로 확보하였다
#### 최근 민영화관련 동향
- 강제력을 가지는일은 국가의 것 그래서 민간경비는 한계가 있다
- 민간경비업과 지역 경찰관서간에 적절한 협력을 위한 협약이 조인되었다
#### 독일의 민간경비시장 구조
- 중소기업, 대기업, 외국계 회사들, 청소대형회사들, 자치단체 단위의 경비조합들
#### 독일 및 유럽의 민간경비시장 현황
- 독일의 민간경호 경비시장은 유럽에서 가장 높은 비중을 차지하고 있다(중소기업 이미지)
#### 경비업의 개념
- 독일영업법 : 영업을 위해 타인의 생명이나 재산에 대한 경비를 하고자 하는 자는 허가를 필요로 한다
- 독일에서는 경비업에 관한 법률이 특별법이 아닌 영업법의 한 조문으로 이루어져 있다
- 경비업에서 경비는 **영업활동의 일환**으로 수행
#### 경비업 관련법규 개관
- 경비업 운영에 대한 관리,감독 책임은 상공회의소에서 담당
#### 경업법 제34조a의 주요내용
- 영업적인 생활 또는 타인의 재산을 경비하고자 하는 자는 관할청의 허가를 필요로 한다
#### 경비업시행령
- 영업법 제34조의a와 경비업시행령에 근거한 경비업에관한일반행정규칙
#### 직업교육 및 자격을 위한 새로운 교육과정
- 독일민간경비아카데미가 이러한 교육프로그램에 대한 책임을 지게 되었다 3년동안 880시간, 코스의 마지막에는 필기시험과 구두시험이 치러진다

## 한국의 민간경비
### 전통적 의미의 민간경비제도
- 고조선부터 1950년 이전까지 공,사 경비형태가 미분화된 상태에서 상주 형태로 경비원을 운용하는 경비활동
### 현대적 의미의 민간경비제도
- 1950~1960 주한미군 군납과 미국의 경제원조에 따른 경비활동 시작부터 오늘날까지
- 민간경비 경영자가 경비활동을 상품으로 하여 경비수요자들에게 영리를 목적으로 전문적인 경비서비스를 제공하던 시기
- 공급자측에서 상품을 개발하여 자체경비와 용역화된 형태의 경비서비스를 제공하는 경비활동
### 전통적 민간경비
#### 고대
- 군사,경찰,민간경비 미분화
- 공경비 조직 : 부족민, 하호
- 사경비 조직 : 부족민, 하호
#### 삼국시대, 통일신라
- 민간경비의 출현 시작, 공경비 위주
- 공경비 조직 : 고구려 = 대모달,말객, 백제 = 위사좌평,5부, 신라 = 서당, 통일신라 = 시위부(군)
- 사경비 조직 : 지방의 실력자들이 해상을 중심으로 사적 경비조직을 활용, 호족들이 자신의 신변보호와 반대세력의 기밀탐지를 위해 사병을 고용함
#### 고려시대
- 다양한 경비조직 출현
- 공경비 조직 : 2군6위, 순마소, 총관, 내군부(장위부, 사위사, 위위부), 성중애마
- 사경비 조직 : 도방, 삼별초, 권력자의 사병, 승병
#### 조선시대
- 다양한 공경비 조직, 민간경비경호조직 상대적 위축
- 공경비 조직 : 내시위, 별시위, 내금위, 내금부, 포도청, 경사복, 의흥(친군위,삼군부), 어영군, 호위청
- 사경비 조직 : 사병(태종이방원은 국왕과 세자의 시위를 제외한 모든 사병을 혁파함), 묘지기, 창고지기, 인삼밭 경비  
#### 구한말
- 최초로 문관ㄱ여찰 시작
- 공경비 조직 : 무위영, 친군용호영, 순사제도, 경비원, 경무청
### 현대적 민간경비
#### 용역경비
> 주한미군에 대한 용역계약에 의한 군납형태의 경비
- 한국은 1960년대 초 미군부대의 용역경비를 담당한 것이 현대적 의미에 있어서 민간경비제도의 효시(시초)라 할 수 있음
- 군납이 부정부패와 비리의 온상이 되자 상공부에 등록한 건설군납, 용역군납, 물품군납만 용역경비 체계를 갖춤
- 계약경비산업이 성장할 수 있는 계기를 마련함
- 1950년 한국전쟁 이후 주한미군에 대한 군납경비를 통해 민간경비산업이 태동하게 되었다

> 미국의 원조에 의한 원조물품
- 1962년 범아실업공사 : 순수한 우리나라 민간경비 시설물에 대한 민간차원의 경비

> 경비 및 검수를 위한 경비
- 경비용역 계약으로 석유류 제품의 인수, 저장 및 유통을 담당하는 KOSCO
- 1958년부터 부산 부두하역장의 화물에 대한 수화물 검수를 해옴
- 1962년 한국석유저장주식회사와 정식으로 용역경비계약을 체결하면서부터 시작
#### 청원경찰 제정(1962년)
- 제3공화국 출범 이후 산업시설, 공장, 금융기관 등 국가 주요 기간산업체의 증가와 북한의 무장게릴라 침투에 따른 한정된 경찰인력을 보조하여 국가 중요시설의 경비를 담당할 목적으로 창설함 -> 1960~70년대에 청원경찰에 의한 국가 주요 기간산업체의 경비가 주류를 이루었음
- 우리나라에만 있는 제도로서 경찰과 민간경비의 과도기적 시기에 만들어진 제도임
- 1962/04/03 청원경찰법 제정, 시행규칙 1967/11/09
#### 용역경비업법 제정(1976년)
- 1976/12/31 용역경비업법이 제정, 당시에는 시설경비업무와 호송경비업무만 도입됨
- 급격한 도시화와 산업화로 범죄가 다시 증가했으나 경찰인력과 장비의 부족, 시국치안 및 대간첩작전의 수행 등으로 대처하는데 한계를 드러냄
- 일본의 직접적인 영향을 받음
- 1999년 3월 기존의 명칭을 경비업법으로 바꿈, 포괄적 개념의 전문경비제도를 도입하는 계기가 됨
- 1977년 설립된 한국경비실업은 경비업 허가 제1호를 취득하였음
- 1989년 용역경비업법은 대통령령으로 정하는 기계경비시설을 설치,폐지,변경한 경우 허가관청에 신고하여야 한다고 규정하였음
#### 한국경비협회 설립
- 사단법인 한국용역경비협회가 1978년 내무부장관의 승인으로 설립됨 ->199년 3월 한국경비협회로 명칭 변경
- 현행 제22조(경비협회)에 규정하고 있다. 제1항은 설립목적, 제2항은 법인성, 제3항은 경비협회의 업무, 제4항은 민법 중 사단법인에 관한 규정을 준용함을 내용으로 함
#### 대기업의 진출과 기계꼉비산업의 발전(1980년대)
- 기계경비는 용어조차 생소하고, 민간경비업체들은 대부분 규모가 영세했기 때문에 경찰이나 군인들이 퇴직 후에 영위할 수 있는 복리사업 정도로 인식되는 분위기였다

> 삼성그룹의 진출과 경제규모 확대 시작
- 1980년대에 들어가면서부터 민간경비산업은 새로운 전환기를 맞이함 영세한 민간경비 시장에 삼성이라는 재벌그룹이 진출하면서 경제규모가 확대되기 시작하였다. 당시 일본에서 민간경비회사들이 성업하고 있다는 점에 자극을 받았기 때문
- 기존업체의 반발과 삼성의 참여방식 변경 : 대기업의 경비시장 진출에 조직적으로 반발하였다
- 1980년 7월 3일 한국경비보장과 지부인수를 통한 경영참여 계약을 맺음과 동시에 일본 최대의 경비회사인 세콤과 합작투자 계약을 체결. 삼성 40%, 일본세콤 40%, 한국경비보장 20% 형태였다
- 에스원의 전신인 한국경비보장은 1981년 상호를 한국안전시스템주식회사로 변경하고 국내 최초로 기계경비사업을 시작하였다, 1996년 주식회사 에스원으로 상호를 변경
- 에스원이 민간경비시장에서 급성장할 수 있었는 이유 중 하나는 삼성그룹 계열사에 대해 경비수요를 적절히 활용
- 에스원은 우리나라 최초로 기계경비를 도입했지만 시장여건이 마련되어 있지 않아서 설립 자본금을 계속 잠식당함
- 국가정책의 변화 :
- 숙직으로 인한 여러 가지 폐단이 사회적 쟁점으로 부상하고 있던중, 은행을 감독하는 재무부는 초기에 은행들의 기계경비의 도입을 금지하는 조치를 취했다. 그러나 그 후 한국안전시스템이 금융권을 시장응로 확보할 수 있었던 것은 재무부가 각 은행에 발송한 시스템경비 절대불가의 공문을 철회하고 자율적으로 입장을 변경하면서부터이다
- 은행권의 기계경비 도입 1983년 조흥은행과 7개 지점을 시범적으로 운영하고 기계경비사업을 계속적으로 확장할 수 있었다
- 에스원의 성공 :경제발전의 효과가 가시화되고 사회적,개인적 부가 확대됨으로써 재산범죄의 위험기회가 상대적으로 증가했고 안전에 대한 사회적 수요가 커졌기 때문에  
- 1980년대 이후 기계경비시스템이 점차 도입되어 민간경비 분야에서 기계경비가 활발하게 시작되었으며, 1980년대에 기계경비가 성장하면서 경비업무의 기계화 및 과학화가 활성화되었다

#### 한국 민간경비 산업의 성장
- 1986년 아시안게임과 1988년 서울 올림픽을 치르면서 경비업계의 발전 -> 1980년대 중반부터 한국의 민간경비산업이 본격적으로 발전하기 시작하였다
- 1993년 국제행사인 대전 엑스포박람회를 민간경비업체에서 행사기간 동안 안전 및 경호경비문제를 자체적으로 별다른 사고 없이 마침 -> 한국 민간경비업체의 능력을 재평가 받았고 민간경비업이 급성장하였다

#### 1990년대
- 1995년 용역경비업법에서 경비지도사가 도입되었고 신변보호업무가 추가되었다
- 1995년 9월 22일 경찰청에서는 용역경비의 담당을 경비국에서 방범국으로 이관했다(청원경찰 : 경비국)
- 우리니라는 경비지도사 시험을 1997년 2월 23일 부터 시행하였다, 2년마다 시행됨
1999년 3월 31일 용역경비업법을 일부개정하면서 경비업법으로 법률명을 변경하였다
#### 2000년대
- 자체경비보다는 계약경비가 발전하고 있다
- 2001년 경비업법으로 기계경비업무가 신고제에서 허가제로 변경
- 인천국제공항의 경비업무를 맡고 있는 특수경비원 제도의 실시 민영교도소  설립 등 국가치안서비스 영역도 분담함
- 2001년 경비업법 4월 7알 전부개정되어 8월 8일부터 기계경비업이 경비업의 한 형태로 제도화 되었음, 특수경비원을 규정하여 청원경찰과 민간경비의 이원화문제가 대두됨
- 2009년 특수경비원 연령상한을 58세에서 60세로 연장하였다
- 2010년 10월 경기도 여주에 최초의 민영교도소인 **아가페 소망교도소**가 개소함
- 2013년 경비업법 개정으로 집단민원현장에 비치된 경비원의 지도,감독 규정이 강화되었음
- 2016년 경비원이 되려는 사람은 대통령령으로 정하는 교육기관에서 미리 신임교육을 받을 수 있다고 규정했다
- 2024년 혼잡,교통유도경비업무가 추가되었다
#### 한국의 민간조사제도
- 영미법계 국가의 자경주의적 전통과 민간조사원제도
- 한국은 정형화된 형식을 갖추고 제도적으로 정착되어 운영되는 것은 아님
- 현행 경비업법에서도 민간조사원제도가 따로 규정되어 있지 않음
- 민간조사와 관련하여 유사한 업무를 하는것은 관할관청에 서비스업으로 신고만 하면 가능함
- 미국니아 일본 등에서 민간조사원제도가 다양한 형태로 이루어짐, 한국도 비슷함
- 민간경비의 한 영역이라고 보기는 어려움
#### 민간조사제도의 장정
- 형사사법기관의 업무부담을 경감시킬 수 있다
- 국민 누구든지 의뢰할 수 있다
- 국가기관의 복잡한 절차를 거치지 않고 피해자는 민간조사기관에 의뢰해서 피해구제를 받을 수 있다
#### 민간조사제도의 단점
- 인권침해의 가능성이 있다
#### 민간경비 산업의 현황과 발전방안
- 2001년 경비업법 개정으로 특수경비원의 총기휴대 허용함 -> 청원경찰의 입지 축소욈
- 기계경비보다 인력경비에 의존성이 높다
- 민간경비의 수요가 특정지역에 편중되어 있다
- 청원경찰과 민간경비의 이원적 운용으로 여러 문제들이 발생함
#### 발전방안
- 경비원에 대한 적정한 임금 보장과 우수인력의 지속적인 확보체제를 구축
- 기계경빕 중심의 민간경비산업의 지향 -> 특수경비업 및 기계경비업의 요건 완화는 오히려 사고위험이 커질 수 있음
- 첨단장비의 개발 : 장비를 현대화하고 첨단장비를 개발하여야 한다
- 특수경비원 쟁의행위 금지 문제의 보완
- 민간경제제도의 단일화 방안이 제기 - 청원경찰의 점진적 축소가 바람직함
- 특수경비업 및 기계경비업 요건 강화
#### 자격증제도
- 경비지도사는 일반경비지도사(시설, 호송, 신변보호, 특수, 혼잡.교통유도), 기계경비지도사(기계경비)로 구분함

> 교육기관 취소
1. 거짓이나 그 밖의 부정한 방법으로 경비지ㅗ사 교육기관의 지정을 받은 경우(무조건 취소)
2. 지정받은 사항을 위반하여 업무를 행한 경우
3. 제11조의 3 제3항에 따른 시정명령을 받고도 정당한 사유 없이 정하여진 기간 이내에 시정하지 아니한 경우
4. 제11조의 3 제4항에 따른 지정기준에 적합하지 아니하게 된 경우(2, 3, 4 재량취소)

> 경비지도사의 직무
- 경비원의 지도,감독,교욱에 관한 계획의 수립,실시 및 그 기록의 유지 - 월 1회 이상
- 행안부령으로 정하는 경비원 교육내용을 2년간 보존하여야 함
- 경비원에 대한 순회점검 및 감독 - 월 1회 이상
- 경찰 및 소방기관과의 연락방법 지도
- 기계경비업무를 위한 긱장치의 운용.감독 - 월 1회 이상
- 오경보방지 등을 위한 기기관리 감독 - 월 1회 이상
- 집단민원현장에 배치된 경비원에 대한 지도.감독

> 경비지도사의 선임.배치기준
- 경비업자는 시.도경찰청의 관할구역별로 경비원 200명 까지는 경비지도사 1명을 선임.배치, 100명 단위로 경비지도사 1명을 추가로 선임
- 30명 이하인 경우에는 경비지도사를 배치하지 않을 수 있다
- 제주특별자치도경찰청과 전라남도경찰청은 경개를 맞닿아 인접한 것으로 본다
- 특수경비원은 일반경비지도사를 선임한다









# 민간경비의 기본유형
## 주체에 따른 분류 - 인력경비와 기계경비
### 인력경비
> 개념
- 개인이나 시설을 각종의 위해로부터 순수한 사람의 능력(오감)을 통해 보호하는 경비형태
### 기계경비
- 관제시설에서 경보를 수신한 때에는 늦어도 25분 이내에는 도착시킬 수 있는 대응체제를 갖추어야 함
- 불법 침입에 대한 감지 및 경고 -> 침입정보의 전달 -> 침입에 대한 대응
## 목적에 따른 분류 - 시설,혼잡,신변,호송,특수
### 계약경비
- 경비상품을 갖춘 경비업자가 경비서비스를 원하는 의뢰인과의 일정한 계약행위를 통해 경비서비스를 제공하는 형태의 서비스
- 개인 및 기관, 기업 등이 중요하다고 판단되는 것을 보호하기 위하여 외부와의 계약을 통해 경비인력 또는 경비시스템을 도입.운영하는 것 -> 현행 경비업법은 계약경비를 전제로 한 것임
### 자체경비
- 조직 스스로가 자체적으로 경비부서를 조직하고 경비활동을 하는 것
- 청원경찰은 자체경비의 일종임
### 경비부서 관리자의 역할
> 경영상의 역할
- 기획, 조직, 채용, 지도, 감독, 혁신

> 관리상의 역할
- 예산과 재정상의 감독, 사무행정
- 경비문제와 관련된 정책수립
- 조직체계와 절차의 발전
- 직원교육.훈련과정의 개발과 다른 고용인들에 대한 경비교육 실시
- 다른부서와 협조와 의사소통의 향상

> 예방상 역할
- 감독, 순찰, 화재와 경비원의 안전
- 교통통제, 출입근지 구역 감시활동
- 경비원들에 대한 규칙적인 감사
- 경비원들에 대한 이해와 능력개발
- 경비장비들 상태점검

> 조사활동
- 관련문서 확인 및 감시, 경비의 명확화, 회사규칙의 위반과 모든 손실에 대한 조사
- 일반 경찰관서와 소방서와의 유대관계 증진과 관련문서의 분류 등

#### 계약경비 장점
- 비용이 저렴함
- 감축, 증원이 원활함
#### 자체경비 장점
- 높은 충성심
- 안정적임
















# 민간경비의 법적 토대
## 헌법상 지위
- 헌법상 직업선택의 자유 및 영업의 자유에 근거함
- 민간경비원의 법적 권한의 근거 : 고용하는 소유자의 권한에서 나옴(일반인)
## 경비업법
- 순수한 개인적 이익(사익)
- 특정고객
- 특별한 권한 없음, 현행범 체포는 할수있음
- 임용권자 : 경비업자
- 감독권자 : 경비지도사, 경비지도사 감독권자 : 경찰청장, 시.도, 관할 경찰관서의 장
## 청원경찰법
- 준공공적 사익
- 기관장 및 시설주
- 경비구역 내에서 경직법에 의한 무기사용 가능, 범죄수사는 못함
- 임용권자 : 청원주
- 감독권자 : 청원주, 관할 경찰서장
### 민.형사법상 지위
- 경비업법상 경비업은 법인만이 영위할 수 있음 -> 벌칙, 과태료, 양벌규정등 규정함
- 경비업법상 경비협회는 민법상 사단법인에 준함
- 민법상 법인 - 사단법인 = 사람, 재단법인 = 물건

### 한반도 정세의 변화
- 북한 핵 문제

- 한국의 노동력 부족
- 이슬람 원리주의 및 분리주의자의 아시아 지역에 대한 태러위협

- 마약범죄의 증가

- 경찰인력 부족

- 총기사용시의 문제점
- 민생치안 부서 근무기피 현상

## 민간경비의 공공관계 개선전략
### 민간경비의 공공관계 중요성
- 공공관계의 개선 : 기업이 공공의 신뢰와 이미지를 증대시키기 위해서 적용되는 모든 관련 정책 및 프로그램
### 공공관계 형성에 영향을 미치는 요소 - 외적으로 전문성을 갖게 하는 인상
- 군대식 복장
- 경비원 고유복장
- 부드러운 이미지의 복장
- 실제적인 전문성 배양 : 적극적이고 긍정적인 태도
### 민간경비와 시민의 관계 개선
- 내부직원 방문객 또는 고객 등 지역사회의 관계자들과 적절하게 상호작용
- 사회적 위치 및 인구사회학적 특성에 따른 공공관계개선 프로그램을 운영
- 민간경비서비스를 이용하지않는 시민이라도 이들은 잠재적 고객이며 적절한 배려 및 긍정적인 이미지 형성은 매우 중요한 일임
- 민간경비원은 정당한 권한 없이 시민의 권리와 자유를 침해하거나 제한하는 일이 있어서는 안 됨
- 경비업체의 영세성과 지역편중으로 인하여 지역사회와 상호협력을 구축하는 것이 필요함
### 민간경비와 언론의 관계 개선
#### 민간경비의 언론매체 대응방법
- 정직성, 전문성, 신뢰성 유지
- 사전적 대응 : 사전에 적극적으로 관련정보를 제공
- 반응적 대응 : 조직에서 침묵으로 일단 사태를 주시하고 공론화되었을 때 정보를 제공
- 무반응적 대응 : 문제가 해결될 때까지 모든 정보를 외부로 노출시키지 않는 방식
#### 효과적인 언론 대응
- 오직 사전적 대응만이 커뮤니티에 대해서 신뢰를 유지.확보할 수 있다는 사실을 인식해야함
- 반응적 대응 : 예상 답변을 만들어 놓고 사태의 전개상황에 따라 대응하는 방식으로 자칫하면 안좋은 이미지를 가져올 수 있음
- 무반응적 대응 : 가장 위험한 대응방식
#### 준수사항
- 정직할 것
- 미소 띤 모습을 유지할 것
- 의무감을 가질 것
- 적절한 답변을 미리 준비할 것
### 민간경비와 경찰의 관계개선
- 민간경비와 경찰의 공공안전과 관련된 전문지식.교육훈련.기술 등의 지속적인 노력
#### 테러위험인물에 대한 정보수집
- 국가정보원장은 테러위험인물에 대하여 출입국.금융거래 및 통신이용 등 간련정보를 수집할 수 있다(출입국관리법, 관세법, 특정 금융거래정보의 보고 및 이용 등에 관한 법률, 통신비밀보호법의 절차에 따른다
- 국가정보원장은 정보 수집 및 분석의 결과 테러에 이용되었거나 이용될 가능성이 있는 금융거래에 대하여 지급정지 등의 조치를 취하도록 금융위원회 위원장에게 요청할 수 있다
- 국가정보원장은 테러위험인물에 대한 개인정보와 위치정보를 요구할 수 있다
- 국가정보원장은 자료를 수집하기 위하여 대테러조사 및 테러위험인물에 대한 추적을 할 수 있다. 이경우 사전 또는 사훙에 대책위원회 위원장에게 보고하여야 한다
#### 제13조(외국인태러전투원에 대한 규제)
- 관계끼관의 장은 내국인.외국인에 대하여 일시 출국금지를 법무부장관에게 요청할 수 있다
- 제1항에 따른 일시 출국금지 기간은 90일로 한다. 다만 계속할 필요가 있다고 판단하는 경우 사유를 명시하여 연장을 요청할 수 있다
- 관계기관의 장은 외국인테러전투원으로 가담한 사람에 대하여 여권법 효력정지 재발급 제한을 외교부장관에게 요청할 수 있다
#### 제14조(신고자 보호 및 포상금)
- 국가는 테러 신고자, 제보하거나 검거활동을 한 사람 또는 그 친족을 보호하여야 한다
- 관계기관의 장은 태러를 사전 예방할 수 있게 하였거나, 체포한사람에 대하여 대통령령으로 정하느 바에 따라 포상금을 지급할 수 있다
#### 제15조(태러피해의 지원)
- 테러 피해를 입은 국민은 관계기관에 즉시 신고하여야 한다. 다만 인질 등 부득이한 사유로 신고할 수 없을 때에는 보호의무가 있는 사람이 즉시 신고하여야 한다
- 국가 또는 지방자치단체는 비용의 전부 또는 일부를 지원할 수 있다
#### 제16조(특별위로금)
- 테러로 생명의 피해를 입은 사람의 유족 또느 장애 및 장기치료가 필요한 사람에 대해서는 그 피해의 정도에 따라 등급을 정하여 특별윌금을 지급할 수 있다
#### 제17조(태러단체 구성죄 등)
- 수괴는 사형.무기 또는 10년 이상의 징역
- 테러를 기획 또는 지휘하는 등 중요한 역할을 맡은 사람은 무기 또는 7년 이상의 징역

## 경비진단활동

- 사전예방조치를 통해 물리적,지리적 위해요소를 파악하고 분석하여 사후복구대책수립이 포함
### 경비계획 수립과정

- 경비진단
- 경비계획수립
- 경비계획서 작성
### 경비요소 조사활동
#### 개념
- 경비위험요소분석절차는 잠재적 위험을 예방.억제하기 위한 경비대응책의 선택과정이고 업무를 수행하는데 있어서 가장 중요한 것 중의 하나이며 위해요소아 경비대상에 대한 다양한 정보를 수집하는 활동임
- 경비진단은 사전조사를 통해 이루어지므로 사전조사란 계약처에서 의 현장청취와 현장조사를 바탕으로 경비대상시설에 잠재하느 경비상의 문제점과 실태를 미리 파악한 후 종합적으로 검토하고 판단하는 일련의 활동
- 사전조사 = 현장청취 + 현장조사 -> 경비대상시설의 이해와 실태를 파악하는 것
1. 현장청취 : 정보 제공받기
2. 현장조사 : 잠재적 위험을 찾는 것
#### 조사대상
- 경비대상물과 주위환경에 대한 구체적이고 세밀한 물리적 요소의 조사
- 모든 집행체제. 절차 등 일상업무에 대한 구체적이고 철저한 조사를 통해 이루어짐
- 경비대상의 현 상태 점검
- 경비방어상의 취약점 확인
- 요구되는 보호의 정도 측정
- 경비활동 전반에 걸친 객관적 분석
- 종합적인 경비프로그램의 수립

> 물리적 착안점

## 경비계획
- 장래 경비업무 수행의 가장 근본이 되는 활동으로 위해분석과 경비요소조사활동에서 나타난 정보와 자료들을 토대로 물리적 요소 등을 고려하여 경비실시의 구체적인 형태를 결정하는 과정
- 경비계획 : 계약처가 요구하는 경비내용을 구체적으로 실시할 방법을 정하는 것
### 필요성
- 건물주들도 거의 무방비로 노출되어 있음 -> 전체적인 경비시스템을 단계적이고 깊은 경비계획이 선행되어야 함
- 내부이용자들을 만족시키기 위해서 절도나 기물파괴 등을 목적으로 하는 내부 침입자를 예빵
- 경비계획은 시간적 측면에서 장.단기적 측면을 모두 고려, 내용적 측면에서 단순한 개념에서 복잡한 개념까지 염두에 둘 필요가 있음
- 경비부서 및 회사 전반적인 상황파악
### 경비계획수립 기본원칙
- 경비구역과 건물출입구 수는 최소한으로
- 4미터 이내에는 강화유리 센서 등 안전장치를 설치
- 건물외부의 틈으로 접근이 가능하거나 탈출이 가능한 지점 및 경계구역으로 보호되어야 함
- 잠금장치는 정교하고 파손이 어렵게 만들어져야 하고, 열쇠 분실시 적절한 조치가 되어야함
- 물건을 선적하거나 수령하는 지역은 분리되어야 함
- 비상시에만 사용하는 외부출입구 등에는 경보장치가 설치되어야 함
- 항구.부두지역 등은 운전자가 차량을 이용하여 물건을 곧바로 창고지역으로 옮기지 못하도록 하고 경비원이 물건의 선적이나 하차를 확인할 수 있도록 설계되어야 함
- 경비관리실은 통행이 많은 곳에 설치하고 직원의 출입구는 주차장으로부터 떨어진 곳에 위치해야함
- 경비원 대기실은 출입구와 비상구에 인접. 격리되어 있는 출입구에 대해서는 경보장치를 설치
- 경비진단결과 나타나는 손실발생 가능성을 고려해야 한다
#### 경비계획의 기본방향
- 관련 분야나 계층의 의견수렴이 이루어져야 하고 실제적용과의 괴리를 최소화하고 절차상의 민주성을 담보하여야 함
- 재정적인 측면 고려
#### 경비계획서
- 계약처가 요구하는 경비내용을 구체적으로 실시할 방법을 정하여 서면으로 작성한 것
- 주된 경비목적과 내용은 일반적으로 도난.화재예방.발견, 출입관리 등임
- 경비회사의 한계를 넘는 내용이나 불가능한 내용, 경찰 등의 공적기관이 아니면 실현할 수 없는 내용은 경비회사의 임무로 정하지 말 것
- 배치포스트 : 경비구역의 블록 및 구역을 설정하는 기본단위
- 포스트 : 인력경비를 수행하는데 있어서 각 경비대상구역이나 임무별로 그 명칭을 붙이고 수량이나 시간 그리고 배치장소를 나타내는 등의 관리단위
- 경비원의 구성 및 배치 : 복수의 경비원이 배치되는 경우 계층제의 원리에 따라 명령 지휘의 체계를 확립하고 책임과 권한의 한계를 분명히 해야 함
#### 일반시설물 경비계획
> 낡은 시설물
- 낡은 장비들은 범죄자들에게 아주 좋은 표적이 되기 쉬우므로 경비계획을 수립하는 데 복잡하고 어려움이 많이 발생하게 됨

> 새로운 시설물
- 현대식 건물양식은 안전요소를 고려하여 설계되는 경우가 많으므로 기본적 경비조사를 실시하고 특수성에 따라 보다 전문적으로 경비계획을 수립하는 것이 필요함
#### 경비실시 형태(수준)
- 경비계획을 수립할 때 조직의 목표.재무상태.경비진단결과 나타난 손실발생 가능성과 정도 등을 고려하여 결정하여야 함
- 모든 시설물마다 똑같이 표준화된 인력경비와 기계경비시스템을 적용하는 것은 아님

##### 실시방식
> 1차원적 경비
- 가장 단순한 경비형태
- 경비원과 같은 단일 예방체제에 의존하는 경비형태

> 단편적 경비
- 경비실시가 필요할 때마다 단편적으로 손실예방 등의 역할을 수행하기 위해 경비조직을 추가해 나가는 경비형태
- 포괄적이고 전체적인 계획이 없음

> 반응적 경비
- 단지 특정한 손실이 발생하는 사건에만 대응하는 경비형태

> 종합적 경비
- 특정한 위해요소와 관계없이 언제 발생할지도 모르는 상황에 대비하여 인력경비와 기계경비를 종합한 표준화된 경비형태
##### 경비수준
- 경비계획 수립을 위한 중요요소
1. 외부출입통제 = 제1차 보호시스템 : 원하지 않는 방문객을 막기 위한 것
2. 내부출입통제 = 제2차 보호시스템 : 내부 침입이나 절도 기타의 도난 등을 막기 위한 것
3. 시설 자체가 지니는 가치
- 경비수준 검도(중요도에 따른 분류)

> 최저수준 경비(레벨1)
- **일정한 패턴이 없는** 외부의 행동을 방어하기 위해 마련된 것
- 외부로부터의 단순한 침입에서부터 무장공격에 이르기까지 모두 방어대상임
- 보통 출입문과 좌물쇠를 갖춘 창문과 같은 단순한 물리적 방벽으로 구성됨
- 미국의 경우 일반가정

> 하위수준경비(레벨2)
- 일정한 패턴이 없는 외부의 행동을 방어하고 **탐지**할 수 있도록 계획된 체계
- 일단 단순한 물리적 장벽과 자물쇠가 설치, 보강된 출입문, 보다 복잡한 수준의 자물쇠, 조명시스템, 기본적 경보시스템, 기본적 안전 장벽
- 작은 소매상점 저장 창고 등

> 중간수준경비(레벨3)
- **대부분의 패턴이 없는 외부행동**과 **일정한 패턴 없는 내부의 행동**을 발견, 방어하도록 계획된 것
- 단순한 기계 좀도두구에서 파괴행위에까지 이름
- 경비체제가 중간수준에까지 이르게 되면 사전에 최저수준과 하위수준에 대해서는 조치가 가능하게 됨
- 불법적인 외부침입과 일부 내부침입을 방해,탐지,사정할 수 있도록 계획되어진 보다 발전되 원거리 경보시스템. 보다 높은 수준의 물리적 장벽, 기본적 의사소통 장비를 갖춘 경비원 등이 동원됨
- 큰 물품창고, 제조공장, 대형 소매점 등

> 상위수준경비(레벨4)
- **대부분의 패턴이 없는(일정한 형식이 없는) 외부 및 내부**의 행동을 발견.저지.방어하도록 계획되어짐
- 위에서 언급한 경비수준이 결합된 보다 높은 수준의 경비체제
- 일정한 형식이 없는 외부와 내부의 이상행동을 감지하여 저지.방어하기 위한 **첨단시스템 장치**를 구비, 높은 의사소통 장비를 갖춘 **고도의 훈련받은 무장경비원**, 장벽 주위에서 감시, **고도**의 경계경보 시스템, **고도**의 조명시스템, 보호장소에서 사고발생시 경비원의 대응과 경찰의 협력이 이루어지는 시스템. 지방 법집행당국과의 조정계획수립 등이 이루어짐
- 교도소, 제약회사, 전자회사 등

>최고수준경비(레벨5)
- **전혀 패턴이 없는** 외부 및 내부의 활동을 발견하여 억제하고 문제를 해결하도록 계획됨
- 레벨1~4까지 모든 계획이 결합되고 **최첨단 경보시스템과 현장에서 즉시 대응할 수 있는 24시간 무장체계**가 갖추어짐
- 핵 시설물 **중요** 교도소, 중요 군사시설, **정부의 특별연구기관**, **일부 외국대사관** 등
##### 경비계획 담당자의 구비조건
- 현재상황을 정확히 파악
- 경비업무에 전문적 지식
- 융통성 있는 창의력.결단력.판단력
- 심리적 안정감, 원만한 대인관계
- 조직의 특유한 풍터나 기질.전통문화를 이해하고 있을 것

## 경비조직화
#### 경비계획 수립 후
- 인력경비의 경우 경비실시를 위해 필요한 조직체계를 만들고, 경비원을 채용하고, 이들 경비원을 교육.훈련하여 경비대를 만드는 것
#### 기계경비의 경우 시스템을 구축하는 것
- 기능별로 조직화 하기 위해 각 부서의 특성을 최대한 살리고 부서간의 역할과 기능을 최대한 보장할 수 있도록 경비시스템이 구축되어야 함 -> 민간경비부서를 독립적으로 설치하지 않고 다른 관리부서와 연계시켜 통합적으로 설치하게 되면 전문성은 저하됨
#### 관한의 위임
- 개념 : 상급자가 하급자에게 권한의 일부를 이전하고 하급자의 권한과 책임하에 임무를 실행하는 것
- 권한의 일부만 이전되며, 중요부분이나 본질적인 임무는 이전할 수 없음
- 상급자는 지휘.감독책임이 있으므로 전적으로 책임이 면제되는 것은 아님
- 중간관리자가 경비업무를 감독할 수 있는 능력이 있다고 판단되면 경비책임자는 중간관리자에게 책임의 범위 내에서 업무를 수행할 수 있도록 권한(재량권)을 부분적으로 위임해야 함
#### 통솔의 범위
- 한 사람의 관리자가 직접 통솔할 수 있는 부하의 합리적인(최대한의) 수
- 최근 부각되는 구조조정의 문제와 관련성이 깊은 것은 통솔범위의 원리임, 계층의 수, 업무의 단순성, 시간적.공간적 요인, 부하의 능력, 감독자의 리더십에 따라 달라짐, 직급조정과 인력재배치 작업으로 연결되는 것임
- 주릐력의 범위와 인간의 능력 간에는 한계가 있다는 것을 전제로 관리능력을 극대화시킬 수 있는 통솔범위를 찾는데 의도가 있음
- 작은 통솔범위를 강조하여 모든 부하직원들과의 밀접한 상호관계를 유지함과 동시에 통제할 수 있기 때문
#### 통솔범위의 결정요인
> 확대요인과 축소요인으로 나뉨
- 시간적 요인
- 공간적 요인
- 직무의 성질(관련업무의 복잡성)
- 인적 요인
- 기술적 요인(정보통신기술)
- 계층제
- 경비조직의 통솔범위
#### 경비 인력의 수요
- 필요한 경비요원의 수요는 해당 시설물의 규모에 비례
- 경비대상조직의 경비인력수요는 경비위해요소조사와 위해분석이 면밀히 이루어진 후 가능함
#### 기타
- 대상조직의 위험, 규모, 물리적인 환경,예산,경비요구에 대한 반응 등을 종합적으로 고려함
- 민간경비 조직화 과정에서 위험성, 돌발성, 기동성 등 경비업무의 특수성을 고려해야 함

> 민간경비 업무의 특수성
- 위험성 : 보호대사아에 대한 인적.물적 공격성을 가지고 있음, 항상 위험성의 내포
- 돌발성 : 공격이 예측 가능한 것이 아니고 대부분 불시에 이루어짐
- 기동성과 조직성 : 공격자 입장에서는 치밀한 계획하에 신속하게 조직을 갖추어 공격하고자 할 것이고 방어자 입장에서는 정반대로 평상시 다른 부서의 내부직원 및 보호대상과 유기적으로 협력하여 사전에 계획적으로 준비하려고 한 것임
















## 민간경비조직의 운영원리

# 위기관리 요약 (민간경비론 230~233)

## 1. 보험과 민간경비의 차이

### 보험
- 화재·범죄로 사망하거나 회복이 어려운 부상을 당했을 때  
  **사후에 금전적 보상만 제공**
- 사전예방 측면에서는 의미가 적음

### 민간경비 ★
- 보호대상이 가진 위험을 **사전에 예방**하려는 활동
- 위험이 발생해도 보험처럼 보상이 이루어지는 것은 아님

### 결론 ★
- 개인이나 조직은 보험과 민간경비를 적절히 활용하여  
  **사전예방 + 사후대응 방법을 모색해야 함**

---

## 2. 위험관리의 기본과정 ★

1. 위험요소 확인  
2. 위험요소 분석  
3. 우선순위 설정  
4. 위험요소 감소  
5. 안전성·보안성 감사·평가  
6. 환류(feedback)

---

## 3. 위험요소 확인

### 위험(risk)의 의미 ★
- 인적·물적 대상이 육체적·경제적·정신적 손해에  
  **노출될 가능성**

### 보호가치의 상이성 ★
- 모든 자산이 동일하게 중요한 것은 아님  
→ 보호 우선순위를 정해야 함

---

## 4. 위험요소 분석 ★

- 자산은 동일한 보호가치를 가지지 않음  
→ 동일한 보호전략이 필요하지 않음

- 자산 유형에 따라 위험요소 분석 필요  
→ 우선순위 설정에 도움

- 위험상황은 현재 상황 중심으로 인식 ★  
→ 과거 자료만으로 계획 세우는 것은 한계 있음

### 위험발생 가능성 분석 ★
- 보호대상별로 낮은 가능성(possibility)과  
  높은 가능성(probability)을 반복 검토해야 함

---

## 5. 위험요소 분석 시 3가지 고려요인 (231)

1. 기업에 영향 큰 높은 수준의 위험발생 가능성 파악  
2. 피해 가능성이 높은 자산 확인  
3. 보호대상의 지역적·공간적 범위와 위치 확인  

---

## 6. 우선순위 설정 ★

### 필요성
- 예산범위 내에서 위험발생 가능성을 최소화해야 함
- 보호가치에 따라 우선순위를 차별 설정

### 어려움 ★
- 사고 발생 시점·장소 확신 불가  
→ 가능성이 높은 것부터 우선순위 설정

### 결정요소 ★
1. 위험발생 빈도보다 **발생결과(악영향)**가 더 중요  
2. 상대적으로 위험 가능성이 높은 대상 우선  
3. 조직 운영에 가장 큰 악영향을 주는 자산이 최우선

---

## 7. 위험요소 감소

- 정책·절차 활용
- 직원 협력관계 구축
- 정형화된 지침서 개발  
→ 조직 성과 향상

---

## 8. 안전성·보안성 감사·평가 (232) ★

### 의미
- 시설의 물리적 안전·보안상태 점검
- 경비조명·경보장치 정상작동 확인
- 프로그램 실행·유지상태 파악
- 정책·절차 준수 여부 확인

### 정형화 필요 ★
- 감사·평가는 일정한 절차 속에서 실시  
→ SOP(표준운영절차) 개발 필요

### 정기적 실시 및 기록 ★
- 정기 점검 + 기록·보존이 중요

---

## 9. 확인된 위험의 대응방법 (233)

1. 위험의 제거 ★  
2. 위험의 회피  
3. 위험의 감소  
4. 위험의 분산  
5. 위험의 대체(보험 등)  
6. 위험의 감수  

### 접근방법 통합
- 대응방법은 명확히 구분되지 않으므로  
  통합적 관점에서 대응책 개발이 최상

---

## ★시험 핵심 암기

- 민간경비 = 위험 **사전예방**
- 과정 = 확인 → 분석 → 우선순위 → 감소 → 감사·평가 → 환류
- 우선순위 기준 = 빈도보다 **결과(악영향)**
- 감사·평가 = **정기적 실시 + 기록**
- 대응 = 제거 / 회피 / 감소 / 분산 / 대체 / 감수















# 📌 민간경비의 채용·교육훈련·윤리 (236~248쪽 합본요약)

---

# 1. 기본논의 (인사관리의 변화)

- 적절한 인사관리를 위해서는  
  과학적 기술과 절차를 활용해 **인적자원을 체계적으로 관리**해야 한다.

- 과거처럼 부적격자를 배제하는 소극적 방식이 아니라  
  오늘날에는 **유능한 인재를 적극적으로 발굴·채용**하는 것이 중요하다.

- 인사관리 체계는 인력수급계획, 신규채용, 교육훈련, 근무성적, 보수, 직원윤리 등  
  여러 요소가 상호 연관되어 있다.

---

# 2. 민간경비원의 채용

---

## 2-1. 신규채용의 의미

- 신규채용이란 관련 분야에서 일할 사람을 새로 뽑는 과정이다.
- 채용은 **모집과 선발과정**으로 구성된다.
- 민간경비 서비스의 질을 결정하는 중요한 요소는 경비원의 **자질**이다.

---

## 2-2. 신규채용의 의의와 중요성

- 민간경비가 양질의 치안서비스를 제공하지 못하는 근본적 이유는  
  ‘민간경비의 약점’ 문제로 인식된다.

- 민간경비는 영리를 추구하지만  
  동시에 공공성이 강한 치안활동을 수행해야 한다.

- 사회적으로 다른 직업을 찾기 어려운 낮은 수준의 인적자원이  
  경비직종에 유입되는 경우가 많다.

- 민간경비의 약점을 끊는 첫 단계는  
  전문지식과 자격을 갖춘 **우수 인적자원을 채용**하는 것이다.

- 계약경비원이 높은 이직률을 보이는 이유는  
  **임금·근무조건 문제** 때문이다.

---

# 3. 민간경비원의 지원자격과 결격사유

---

## 3-1. 일반경비원 결격사유

다음에 해당하면 경비원이 될 수 없다:

- 18세 미만
- 피성년후견인
- 파산선고 후 복권되지 않은 자
- 금고 이상의 형을 받고 집행이 끝나거나 면제된 날부터 5년이 지나지 않은 자
- 금고 이상의 형의 집행유예 기간 중인 자
- 벌금형을 선고받고 10년이 지나지 않은 특정 범죄자 등

---

## 3-2. 특수경비원 결격사유

- 18세 미만 또는 60세 이상
- 심신상실자, 마약·알코올 중독자 등 정신적 제약이 있는 자

---

## 3-3. 청원경찰 결격사유

- 금고 이상의 형을 받고 일정 기간이 지나지 않은 자
- 벌금형(300만원 이상 등)을 받고 일정 기간이 지나지 않은 자
- 미성년자 대상 성범죄 등 특정범죄 전력자 등

---

# 4. 경비업자의 범죄경력조회 요청

- 경비업자는 경비원을 선발·채용·배치할 때  
  결격사유 해당 여부 확인을 위해 경찰서장에게 범죄경력조회를 요청할 수 있다.

- 경찰은 결과를 통보해야 한다.

---

# 5. 민간경비원 채용상의 문제

---

## 5-1. 형식적인 채용절차의 문제

- 선발 과정이 형식적으로 흐르고  
  절차를 무시하는 경우가 있다.

- 외형적 신체조건만 기준으로 뽑는 경향도 있다.

- 최저임금 지급 등 근무조건이 열악해 인력확보가 어렵다.

---

## 5-2. 무자격자에 의한 업무수행 문제

- 경비업자는 경비업무에 종사할 수 없는 자를 배치하면 안 된다.
- 신임교육을 이수한 사람을 배치해야 한다.
- 위반 시 배치폐지 명령 가능

---

# 6. 민간경비의 교육훈련

---

## 6-1. 교육훈련의 의미

- 직무수행에 필요한 지식·기술을 습득하고  
  능력을 개발하는 계획적 활동이다.

- 단기적 효율성만이 아니라  
  장기적·포괄적 발전을 목표로 한다.

---

## 6-2. 교육훈련의 필요성

- 사회 변화와 범죄 변화에 대응하기 위해 정기적 교육 필요
- 민간경비의 약순환(낮은 인지도 → 인재유입 부족 → 교육투자 감소)을 끊어야 한다.

---

## 6-3. 교육훈련의 효과

- 조직 목표(생산성) 달성에 기여
- 경비원의 사기와 경력발전 촉진
- 전문성 함양
- 조직통제·조정문제 감소
- 업무상 실수·낭비 감소
- 조직 안정성과 융통성 확보
- 민간경비산업 발전

---

## 6-4. 경비업법 교육 규정

- 경비업자는 경비원에게 신임교육 및 직무교육을 받게 해야 한다.
- 특수경비원은 신임교육 + 정기 직무교육 필수

## 8-4. 시민의 인식전환

# ✅ 시설경비활동 요약정리 (262~271쪽)

---

## 1. 경비조명 (262~263)

### (1) 조명의 종류

- **상시조명(continuous lighting)**
  - 항상 켜져 있는 조명
  - 주로 **창문, 벽의 외곽**을 비추는 데 사용
  - 장점: 주변 접근 감시 + 침입 예방 효과

- **예비조명(standby lighting)**
  - 상시조명을 **보충하거나 예비용**으로 설치
  - 수동 또는 자동으로 작동 가능
  - 우범지역·취약지역에서 유용

- **이동조명(movable lighting)**
  - 휴대 가능한 조명장치
  - **일시적으로 필요한 특정지역**에 설치 가능
  - 상시조명 또는 예비조명의 보충 역할

- **비상조명(emergency lighting)**
  - 정전, 긴급상황 시 사용하는 조명
  - 평상시 정상작동 여부를 **점검해야 함**
  - 이상 발견 시 즉시 원상복구 필요

---

### (2) 관리·유지

- 조명설비는 **정기적인 점검**이 필요
- 전기회로 이상, 합선 → **화재 예방** 목적

---

### (3) 설치 원칙

- 경비조명은 사고발생지역을 **직접 적절히 비춰야 함**
- 조명이 경비원의 시야를 방해하면 안 됨
- 그림자·사각지대가 생기지 않도록 해야 함
- 조명으로 인해 주변 사람에게 불편을 주면 안 됨

---

### (4) 출입·통로 조명

- 출입구·내부통로 조명은 충분히 설치해야 함
- 보행자 출입구는 **문 앞 약 7m까지 비춰야 함**
- 차량 출입구는 보행자 출입구보다 **약 2배 정도 넓게** 비춰야 함

---

### (5) 기타

- 경계구역 접근을 제지하기 위해 체계적인 조명계획 필요
- 통로지역에 설치된 조명시스템은 경비프로그램의 필수 요소

---

---

## 2. 시설 내부경비의 의의와 중요성 (263)

### (1) 의의

- 시설경비는 단순히 외부경비만이 아니라 **내부경비도 포함**
- 창문, 출입문, 서류·물품 안전관리 등이 포함됨

### (2) 중요성

- 내부경비의 목적은 시설 내 침입·절도·훼손 등을 예방하는 것
- 시설마다 보호대상 중요성이 달라 경비방법도 달라짐

---

---

## 3. 창문경비 (263~264)

### (1) 창문이 중요한 이유

- 외부침입자가 가장 쉽게 이용하는 통로가 **창문**
- 창문 안전조치는 시설경비에서 매우 중요

---

### (2) 창문 보호 대책

- 창문은 **물리적 대책 + 정기적 확인** 필요
- 강화유리, 외부침입 감지시스템 설치 가능

---

### (3) 방호창문(security glass)

- 일반유리보다 강도가 높아 충격에 강함
- 단점: 가격이 비싸고 설치가 어려움

---

### (4) 창문파괴 대비책

- 창살 설치
  - 장점: 침입 가능성 감소
  - 단점: 미관 문제, 비상탈출 곤란, 부식가능

- 창문 경보장치 설치
  - 유리파괴 센서(충격·진동 감지)
  - 침입 시 경보 발생

---

---

## 4. 출입문 경비 (264~266)

### (1) 출입문 경비요소

1. 사람과 수하물 출입통제  
2. 차량 출입통제  
3. 출입문의 물리적 안전성  
4. 출입문 잠금상태 점검  
5. 중요도에 따른 출입구 구분  
6. 비상구 확인 및 구별시설 설치  

---

### (2) 통행절차

- 내부직원과 외부방문객을 구분해야 함
- 방문증·사원증 패용 등 신분확인 절차 필요
- 차량 출입도 출입증 발급 및 주차구역 지정 필요

---

### (3) 비상구 관리 ★

- 비상구는 항상 활용 가능해야 함
- 비상등과 비상벨 설치 필요
- 비상구를 통한 외부 침입 차단 장치 필요

---

### (4) 중요지역 출입문 관리

- 통신기밀실, 전산실, 연구실 등은 별도 보안통제 필요
- 예측 가능한 출입인원을 최소화하고 보안체계 강화

---

## 5. 출입문 잠금관리 (266)

# ✅ 시설경비활동 요약 (272~281쪽) ★포인트 포함

---

## ★ 기출/암기 포인트(별표)

- ★ 방문객 통제: 신원확인 + 대기실 절차 + 신분증 부착
- ★ 직원 통제: 신분증 발급, 최근사진, 직위별 출입구역 차등
- ★ 화물 통제: 무단반출 시 조사 후 반드시 보고
- ★ 차량 통제: 용도별 출입허가증 + 주차지역 제한
- ★ 경보시스템 의의: 사고 시 신호전달, 물리적 방어+경보+CCTV 필요
- ★ 순찰활동: 취약구역 집중, 긴급대응체계 필요
- ★ 중앙모니터 시스템 장점: 오경보↓, 신속대응
- ★ 상주경비 시스템: 대응最快 but 인력 많이 필요
- ★ 감지기 설치 고려사항: 오작동/오경보 방지, 비용효과 고려
- ★ 광전자식(레이저) 센서: 넓은 감시 가능 but 벌레·동물 오보↑
- ★ CCTV: 안내판 설치 의무(개인정보보호법)
- ★ 순찰경비 = 선·면 경비 / 출입관리 = 점 경비

---

# 1. 시설물 내부 출입통제 (272쪽)

## (1) 방문객 통제 ★
- 방문객은 반드시 **신원확인 절차**
- 자주 출입 시 검색시스템 설치
- 방문예정 없이 방문 → **대기실**에서 절차 진행
- 방문객은 **신분증 부착**
- 중요지역 이동 시 철저히 확인·감시
- 시설 중요도에 따라 이동구역 제한

---

## (2) 직원 통제 ★
- 내부직원이 많으면 개별 확인 불가 → **신분증 발급**
- 신분증 조건:
  - 최근(2년 이내) 사진 부착
  - 직위·직급별 출입구역 차등

---

## (3) 화물 통제 ★
- 반출입 물건 확인 절차 필요
- 허가 없는 반출 → 조사 후 반드시 보고
- 통제규정은 내부직원에게 충분히 숙지

---

# 2. 차량 통제 (273쪽) ★

- 내부직원이 자료·물건 차량 반출 가능 → 차량통제 중요

방법:
1. 차량 용도별 출입허가증 발급
2. 신분확인/주차스티커 활용
3. 차량별 주차 가능 지역 제한

---

# 3. 경보·감시 시스템 의의 (273쪽) ★

- 사고 발생 시 신호 전달 → 대응 가능
- 안전조치해도 침입 가능성 존재
→ **물리적 방어 + 경보장치 + CCTV 필요**

구성:
- 물리적 방어: 침입시간 지연
- 경보장치: 침입자 감지·전달

---

# 4. 순찰활동 (273쪽) ★

- 내부직원 또는 경비원이 사후 점검
- 범죄 가능성 높은 곳 집중 확인
- 일정 경로 유지하며 정기 순찰
- 위험상황 대비 **긴급대응체계 필요**

---

# 5. 경보시스템 종류 (273~274쪽)

## 침입경보시스템
- 외부 침입 예상 통로·창문에 센서 설치
- 불법침입 시 경보 발령

## 화재경보시스템 ★
- 화재 위험 높은 곳 설치
- 소화장치와 연계 가능

## 특수경보시스템
- 실내온도 이상, 기계고장 감지 등

---

# 6. 경보체계 유형 (274쪽) ★

## 중앙모니터 시스템(중앙통제) ★
- 감지기/CCTV 설치 후 중앙에서 원격감시
- 장점:
  - 경비원이 직접 확인 → 대응 적절
  - 오경보·오작동 우려 낮음

---

## 상주경비 시스템 ★
- 침입 예상 지점마다 경비원 배치
- 장점: 가장 빠른 대응
- 단점: 인력 많이 필요

---

## 제한경보 시스템
- 사이렌·비상등으로 직접 전달
- 단점: 사람이 없으면 무용

---

## 외래지원 경보시스템 ★
- 경찰서·소방서 자동 연락
- 단점: 오경보 문제 큼

---

## 다이얼 경보시스템 ★
- 침입 시 112
- 화재 시 119 자동 연락

---

## 국부적 경보시스템
- 일부 구역만 경보

## 로컬 경보시스템
- 경비원이 현장 출동 대응

## 광전자식 센서(광감지기) ★
- 레이저 차단 시 경보

- 단점: 민감 → 오경보 가능

## 진동탐지기
- 금고 등 보호

## 열감지기
- 침입자 체온 감지
- 단점: 온도 변화 민감

## 적외선 감지기 ★
- 적외선 차단 시 경보
- 단점: 햇빛·헤드라이트 오작동 가능

## 전자파 울타리
- 전자벽 형성, 오보↑

## 고주파 울타리(마이크로웨이브)
- 감지거리 20~200m
- 동물 오보 많음

단점:
- 범행이 다른 장소로 이동 가능

★ 개인정보보호법:

# ✅ 시설경비활동 요약 (282~286쪽) ★포인트 포함

## ★ 기출/암기 포인트(별표 표시)

- ★ 단독순찰 단점: 다수 범법자 대응 어려움, 위험 많음
- ★ 도보순찰 장점: 상세관찰 가능, 사고 시 신속대응
- ★ 순찰근무자의 준수사항: 단정한 복장, 장구점검, 정숙근무
- ★ CCTV 설치는 개인정보보호법 제25조 적용(공개장소 원칙적 금지)
- ★ CCTV 설치 시 안내판 설치 + 의견수렴 + 운영방침 마련 필수

- 은밀하게 주의를 집중하며 순찰 가능

❌ 단점 ★  
- 다수 범법자 대응 어려움  
- 불의 공격에 대한 위험 많음

### (2) 복수순찰
- 경비원 2명 이상 순찰

❌ 단점  
- 많은 경비인력 요구

✅ 장점  
- 범죄·사고 시 즉각적 충동조치 가능  
- 유사시 비상연락체계 가동 가능

## 2. 순찰수단(기동성)에 따른 구분 (282쪽)

- 비교적 상세 관찰 가능

### (3) 자동차순찰
- 순찰차·오토바이·자동차 이용

✅ 장점  
- 광대지역을 짧은 시간에 순찰 가능

## 3. 순찰지역 구분 (282쪽)

## 4. 효과적인 순찰방법 (282쪽)

- 정신순찰 + 난선순찰 혼합

## 5. 순찰의 원칙 (282~283쪽)

1단계: 조기발견·처리  
2단계: 기록과 보고

4단계: 감소대책 수립

## 6. 순찰근무자의 준수사항 (283쪽) ★

## 8. 건물 내부 순찰 점검사항 (283쪽)

## 9. 건물 외부 순찰 점검사항 (284쪽)

- 출입문·창문 잠금 점검
- 배수구 점검
- 옥외 조명 상태 확인

# 제2절 화재예방경비 (p.297~304) ★복붙요약

---

## 1. 화재의 개념 ★
- 화재: 인간이 의도하지 않거나(또는 고의로) 불이 난 재해
- 화학적 폭발현상을 수반하기도 함
- 불 = 가연물(착화물) + 산소(공기) + 점화원 반응 → 열·빛 발생

---

## 2. ★화재원인(=화원·발화)의 3요소
1) **화원(불씨·열)**
2) **가연물(착화물·연료)**
3) **공기(산소, 산화공급원)**

### ★핵심
- 3요소 중 **하나만 제거해도 화재는 진화**

---

## 3. 가연물의 종류
- 기체 가연물: 밀폐용기 보관(확산 우려)
- 액체 가연물: 휘발유·신나 등(증발연소)
- 고체 가연물: 목재·종이 등(표면연소)

---

## 4. 소화방법 4가지 ★
1) **냉각소화**: 물·CO₂로 온도↓
2) **질식소화**: 산소 차단(포말·모래)
3) **제거소화**: 연료(가연물) 제거
4) **억제소화(부촉매소화)**: 연쇄반응 억제(할론계)

---

## 5. 화재의 특성 ★
- 열·화염 외에 **연기·유독가스** 발생 → 사망 원인 多
- 연기 상승속도: 수직 3~4m/s, 수평 0.3~0.8m/s
- 공기 중 산소 약 21%
  - 산소농도 **16% 이하** → 연소 중단

---

## 6. ★화재의 종류(A~E급) (한국은 A,B,C,D,E)
### A급(일반화재)
- 목재·종이·섬유
- 소화: **냉각소화(물)**

### B급(유류화재)
- 휘발유·알코올·기름
- 소화: **질식소화(포말, CO₂)**

### C급(전기화재)
- 전기설비(감전 위험)
- 소화: CO₂, 분말 (물 사용 금지)

### D급(금속화재)
- 마그네슘·나트륨 등
- 소화: **건조사·마른모래**
- ★물·수분 소화제 사용 금지

### E급(가스화재)
- LPG·LNG
- 소화: **가스밸브 차단 후 냉각**

---

## 7. ★화재경보시스템(감지단계)
### (1) 초기(감지)단계
- 연기·불꽃은 안 보이고 **감지만 가능**
- 감지기: **이온식, 광전식 연기감지기**

### (2) 그을음 단계
- 연기 보이지만 불꽃 없음

### (3) 불꽃발화 단계
- 불꽃·연기 보임
- 감지기: **적외선 감지기**

### (4) 열단계(열감지단계)
- 강한 고열 감지
- 감지기: 정온식, 변온식

---

## 8. 소화기 종류 ★
### 분말소화기(A,B,C급)
- 대부분 화재 사용 가능
- 단점: 분말이 남아 오염

### CO₂소화기
- 질식+냉각 작용
- 전기·유류화재에 적합
- 단점: 질식 위험(밀폐공간 주의)

### 할론소화기
- 소화효과 매우 큼
- 단점: 독성·오존층 파괴 → 현재 생산X

---

## 9. 제거소화 예시 ★
- 전기화재: 전원 차단
- 유류화재: 가연물 제거
- 가스화재: 밸브 차단

---

## 10. 스프링클러(자동확산소화용구)
### 장점 ★
- 신속 대응, 넓은 면적 동시 진압
- 초기진화 효과 큼
- 물 사용 → 경제적

### 단점
- 초기시설비 필요
- 시공 복잡
- 물 피해 발생 가능

---

## 11. 화재대비 안전교육 ★
- 모의훈련, 이론교육, 장비사용법 교육
- 소화기 관리: 습기·직사광선 피함, 수시점검

---

## 12. 화재 발생 시 조치요령 ★
### 화재신고
- 큰 소리·전화(119)로 신속 신고

### 초기소화 주의
- 전기화재: 물 사용 금지
- 유류화재: 물 사용 금지
- 가스화재: 밸브 차단 우선

### 대피요령
- 문 열기 전 손잡이 뜨거우면 열지 말 것
- 낮은 자세로 이동(연기 흡입 방지)
- 엘리베이터 이용 금지 → 계단 이용

---

## 13. ★백드래프트(Backdraft)
- 밀폐공간 화재 진행 중 산소 부족 상태에서
  문 열려 산소 유입 → 폭발적 연소

---

## 14. 피난유도 원칙 ★
- 연기 상승속도 > 사람 계단속도
  → 불가능하면 옥상/구조 가능한 장소로 유도
- 화재층 기준: 상층→하층 순 피난
- 최종대피자: 방화문 닫고 대피

# 제3절 재난관리경비 (p.308~317) ★요약정리

---

## 1. 재난관리경비의 의의 ★
1) 자연적 재난은 예고 없이 발생 → 예방·사후대책 어렵다  
2) 신속한 대응 없으면 인명피해 커짐  
3) **사전 훈련 + 사후 복구대책** 필수  
4) 비상계획은 재난 시 생존기회 증가 목적

---

## 2. 경비부서(경비원)의 역할 ★
- 물리적 경비뿐 아니라 **인적·물적 안전 책임**
- 비상상태 발생 시 **초동조치 책임**
- 질서유지, 대피유도, 화재진압, 응급환자 구조·처치 주도
- 경찰·소방·의료 도착 후에는 **업무 보조**
→ 비상 시 경비원은 **비상요원 역할 수행**

---

## 3. 특정한 위해(위험) ★
- 위험 노출 정도는 시설·지역·국가에 따라 다름
- 화재·폭발 위험은 **화학공장**이 더 큼
- 민간경비는 잠재적 위해를 **제거·감소**시키는 활동

---

# 4. 자연적 재난

---

## (1) 홍수·태풍·산사태 ★
- 태풍과 함께 발생 많음 → 인적·물적·정신적 피해
- 예측 가능 → 침수지역 배수시설·하수구 점검
- 지대 낮으면 물건을 높은 곳으로 이동
- 습기로 붕괴 위험 시 사전 방수조치
- 산사태 대비 사전교육 필요

---

## (2) 지진 ★

### 지진의 종류
1) 단층지진: 지각 변동 → 균열 발생(대규모 지진)
2) 화산지진: 화산폭발로 발생
3) 함락지진: 지층 붕괴로 발생
4) 인공지진: 발파·압축공기 등 인위적 발생

### 지진파 종류 ★
- **P파**: 빠름
- **S파**: 느림, 피해 큼
→ 자연지진은 S파 피해가 더 큼

### 지진 발생 시 대응 ★
1) 지진 후에도 여진 계속 → 방심 금지
2) 다수 사상자 발생 가능 → 부상자 치료 우선
3) 공황·약탈·방화 가능 → 근무태세 최고단계
4) 관계기관 주도하에 협력
5) 가스밸브 잠그고 대피
6) 엘리베이터 안이면 모든 층 버튼 후 정지층에서 하차

---

# 5. 인위적 재난

---

## (1) 건물붕괴 ★
- 산업화 부산물 → 대규모 인명·재산피해
- 경비원은 균열·침하 등 붕괴위험 점검 필요

### 붕괴사고 시 행동요령 ★
1) 침착하게 사람들의 심리 안정 유지
2) 벽·기둥 균열 확인 → 추가 붕괴 대비
3) 엘리베이터 금지, 계단으로 대피
4) 부상자 확인 → 응급조치 + 탈출로 확보
5) 건물 밖 안전지역 대피 후 119 신고
6) 구조 기다릴 때는 체력소모 줄이고 자세 유지

---

## (2) 항공기 추락 ★
- 이용 증가로 사고 가능성 증가
- 사고는 사망보다 부상자가 많음
- 주요원인(통계)
  - 인적요인(조종사 실수 등) 83%
  - 기체결함 13%
  - 자연재해 4%

### 행동요령
- 침착히 상황파악
- 신고 및 관계기관 협조
- 부상자는 안전지대로 이동 후 응급조치

---

## (3) 가스폭발 ★

### 폭발 3요소 ★
1) 가스누출로 밀폐공간 체류
2) 공기와 혼합해 일정 농도 도달
3) 점화원(불씨) 존재

### 가스누출 시 조치 ★
1) 사람을 건물 밖 안전지대로 대피
2) 문·창문 열어 환기
3) 연소기 점화원 차단 + 중간밸브 잠금
4) 전기스위치·화기 사용 금지
5) LPG는 공기보다 무거워 바닥에 체류 → 낮은 곳 환기

---

## (4) 반달리즘(Vandalism) ★
- 무차별적·반사회적으로 시설·문화재 파괴
- 사전경고 없이 목적 없이 조직적으로 발생
- 발견 즉시 경찰 신고 + 추가범행 방지

# ✅ 시설경비활동 핵심정리 (318~334쪽)

---

## 1. 경찰관에 인계해야 하는 사항 (p.318)

경비원이 사건·사고 또는 범죄 사실을 확인한 경우  
경찰에 사실 그대로 전달해야 한다.

### 전달내용

1. 경비대상시설의 명칭, 업종, 소재지, 책임자, 전화번호  
2. 경비원의 성명 및 경비업체명  
3. 발견한 시간 및 내용  
4. 연락·신고 받은 시간 및 내용  
5. 제3자 연락으로 알게 된 경우 → 발견자의 성명·주소·연락처  
6. 현장보존 조치 범위와 방법  
7. 현장보존 시간  
8. 현장보존을 위해 취한 조치·행동 범위  
9. 조치 전 현장에 있던 사람의 인적사항  
10. 조치 후 출입한 사람의 인적사항  
11. 불가피하게 물건 이동 시 → 이유 + 이동 전 위치  
12. 관계자의 성명·주소·연락처  
13. 관련 정보 제공자의 성명·주소·연락처  

---

## 2. 수상한 자의 의의 (p.318)

### 수상한 자란?

- 시설경비 중 **출입이 허가되지 않은 구역에 들어가려는 자**
- 불특정 다수가 출입하는 시설에서  
  수상한 거동·복장·휴대품 등으로 판단되는 자  
- 주변 환경에 맞지 않는 자

---

## 3. 수상한 자 대응요령 (p.318)

1. 시설관리권에 의거한 자세로 기선을 제압  
2. 침착하고 냉정한 말투 사용 (인권침해 금지)  
3. 상황에 따라 입장 거부·퇴거 요구 조치 필요

---

## 4. 유의사항 (p.318)

1. 상대의 안색, 눈동자 움직임, 침착성 관찰  
2. 냉정한 태도 유지, 상대를 흥분시키지 말 것  
3. 처음부터 상대가 납득하도록 설명  
4. 가능하면 여러 명이 대응  
5. 긴급 시 신속히 경찰기관 신고

## 5. ★ 수상한 자 발견 시 사고방지 (p.318)

4. 야간에는 어두운 장소 피하기  
5. 상황에 따라 **경봉을 즉시 사용할 수 있도록 가까이 둘 것**

# 1. 숙박시설경비

## 의의
- 국제행사·관광 증가 → 숙박시설 이용 증가
- 규모에 맞는 **체계적 안전관리 필요**

## 경비부서 규모 결정요소
- 객실 수, 건물 규모, 직원 수
- 지역 범죄발생률
- 경제성·조직체계 여부
- 계절·기후조건
- 경찰·소방 등 공공기관 지원 가능성

## 특징
- 출입 제한이 거의 없음  
  → **절도, 성매매, 화재, 반달리즘 위험**

## 내·외부 문제
- 투숙객 현금·귀중품 많음 → 사고 소지 큼
- 종업원은 보조키로 객실 접근 가능 → 보안 취약
- 이용객이 장시간 외출 → 객실 내 귀중품 보관

## 이용객 문제
- 이용객이 범죄를 저지르기도 함
- 집기 절도, 시설 파손(반달리즘)

## 화재피해가 큰 이유
- 최초 발견 지연
- 경보기 작동 지연
- 투숙객 신고 늦음
- 내부 마감재가 인화성
- 소방교육·훈련 부족
- 체크인 시 주의사항 설명 부족
- 대피 유도 직원 부족

## 경비원의 임무(순찰)
- 기본임무: 내·외부 순찰
- 순찰 종류: 정상순찰(기본) / 기습·집중순찰
- 점검사항  
  - 로비 혼잡·거동수상자 감시  
  - 시설물 파손 여부  
  - 화재 등 비상사태 대비

4. 주기적(연 1회 이상) 효과 측정
5. 부서별·개인별 경비조사
6. 정기적 범죄예방교육
7. 홍보물·연락망 구축

## 진행 절차
- **계획 → 준비 → 방법활동 → 평가·측정**

## 방법활동
- 범죄 인식
- 위험성 홍보
- 평가
- 강연·훈련

## 평가·측정
- 효과성 측정을 위해 평가 필요
- 정보 수집 후 재분석 과정

## 교육시설 주요 범죄위험요소
- 절도, 폭력, 성폭력, 조직폭력, 약물남용, 반달리즘 등

# 3. 의료시설경비

## 의의
- 의료시설은 위험 노출 + 비상상황 시 대규모 피해 가능

## 안전관리 특징
- 응급실: 항상 개방 → 출입 통제 어려움
- 생명 긴박 상황 많음
- 중환자·감염 위험 존재 → 철저한 출입통제 필요
- 약품창고·수술실 등 보호 필요
- 방문객·환자 출입이 지속적 → 관리 어려움

## 취약성
- 출입 통제가 어렵고 구조가 복잡
- 화재 발생 시 대피 취약(환자 이동 불가)

## 안전확인 점검
- 화재대피시설 점검(소화기·비상구 등)
- 비상구·비상계단 장애물 여부
- 위험지역 표시·부착 상태
- 산소·가스시설 점검
- 주요구역 지속 감시·순찰

## 국가중요시설 대상

- 공항·항만 등 교통시설
- 연구시설, 교정시설 등

## 지정
- 국방부장관이 관계기관장 + 국가정보원장과 협의하여 지정

## 관리
- 시설주는 자체 방호계획 수립·시행
- 경찰서장·지역군사령관은 방호지원계획 수립

## 분류 기준(중요도)
- 시설 기능·역할의 중요성 정도

### 실질적 분류
- **가급**: 광범위 지역에 결정적 영향
- **나급**: 일정 지역에 중대한 영향
- **다급**: 제한된 지역에 상당한 영향

### 형식적 분류
- 행정시설 / 산업시설

# 핵심 암기 포인트

- 숙박시설: **출입자 통제 어려움 + 절도·화재 위험**
- 교육시설: **범죄예방프로그램 절차(계획→준비→활동→평가)**
- 의료시설: **응급실 개방 + 감염·중환자 관리**
- 중요시설: **대통령 지정 / 국방부장관 협의 / 가·나·다급 분류**

# 중요시설경비 핵심 요약

## ★ 방호시설
- 울타리·철책, 초소·망루, 방호벽

## ★ 전자적 공격

---

## ★ 악성코드(멀웨어)
- **정의** : 악의적 소프트웨어
- **유형** : 바이러스, 웜, 트로이목마 등

---

## ★ 바이러스 세대
- **1세대(원시형)** : 구조 단순, 분석·백신 제작 용이
- **2세대(암호화)** : 코드 암호화, 탐지 회피
- **3세대(은폐형)** : 존재 숨김, 백신 회피
- **4세대(갑옷형/다형성)** : 형태 지속 변경
- **5세대(매크로)** : 오피스 매크로 기반 감염

---

## ★ 서비스거부공격(DOS/DDoS)

- **분산서비스거부(DDoS)** : 다수 PC 동시 공격

  - 출력조작

# 컴퓨터안전경비 ★초압축 요약 (표시·별표 기준)

## ★ 패치·포트
- 패치(patch) : 취약점 보완 프로그램
- 보안패치/핫픽스(hotfix) : 보안 취약점 전용 패치
- 모든 통신은 **포트** 통해 이루어짐
- 사용하지 않는 포트는 **반드시 차단**
- 라우터·서버 미사용 포트도 폐쇄

---

## ★ 방화벽(Firewall)
- 내부 네트워크 ↔ 외부 인터넷 사이 **방화벽 필수**
- 정보의 **손실·변조·파괴 최소화**
- 보안정책에 따라 **차단/허용** 결정
- 내부정보 외부 유출 방지
- 외부·내부 방화벽 구분
  - **외부방화벽** : 외부 침입 차단
  - **내부방화벽** : 내부 부서 간 정보 유출 방지

### 방화벽 유형
- **스크리닝 방화벽** : 패킷 필터링, 1세대
- **애플리케이션 게이트웨이** : 패킷 내용 검사, 2세대
- **서킷 레벨 게이트웨이** : TCP 연결 기반, 고급 보안

---

## ★ 내부지침
- 보안 내부지침·정책 문서화
- 무단 소프트웨어 설치 금지
- 이메일 첨부파일 주의
- 암호 사용 지침 필요
- **암호수명(password age)** : 90일 또는 180일 변경
- 퇴직자 계정 즉시 사용 금지

---

## ★ 조사·검사
- 시스템 진단 시 **네트워크 검사 중요**
- 보안 점검 툴 활용
- 정기적 보안 결함 점검

---

## ★ 물리적 통제
- 중요 시설은 안전구역 배치
- 접근경계 명확화
- 잠금장치·출입통제 필수

---

## ★ 직원 관리
- 정보보안 사고 다수는 **내부자**
- 내부직원 절차적 통제 중요
- 신규채용·대상자 보안교육 필수

# 공연·행사장의 안전관리 ★초압축 요약

- 안전사고의 **1차적 책임: 행사주최자**

- 국가·지자체 주관 축제·행사·스포츠경기 등 대상

### ★ 중요성
- 군중심리·군중행동 이해 필수
- 정리·유도·안내 능력 중요

### ★ 대상
- 행사 참가자 + 우연히 유입·통과하는 사람 포함

### ★ 기본 원칙
- **예절 중심 경비**
- 주최자·경찰·소방과 협력
- 질서 있는 군중 이동 유도

# 호송경비업무 ★초압축 요약(해당 페이지만)

## 1. 호송경비의 의의 ★
- **현금·유가증권·귀금속·상품 등 운반 중 도난·화재 위험 방지**
- 근거: 「경비업법」 제2조

## 2. 호송경비업 기준 ★
- **무술유단자 일반경비원 5명 이상**
- **경비지도사 1명 이상**
- **자본금 1억 원 이상**
- **호송차량 1대 이상**
- **현금호송백 1개 이상**

## 3. 호송 방식 ★
- **분리호송**: 운송업자 차량 운송 + 경비업자 경비
- **동승호송**: 운송차량에 경비원 동승(가장 일반적)
- **휴대호송**: 경비원이 직접 휴대(소형·고가물품)
- **편성호송**: 운반·수행·경호조 복합 편성(고위험)

## 4. 호송계획 핵심 ★
### (1) 이동도로 취약요소
- 교차로, 공사구간, 협소도로

### (2) 시간대
- 출발·도착 직전
- 적재·하역 중
- **야간 호송 = 고위험**

### (3) 인적 요소
- 운전자·경비원 신원확인
- 질병·부상 대비

## 5. 호송경로 선정 시 회피 ★
- 상습정체·사고다발 구간
- 교차로 밀집
- 공사구간
- 학교·시장 등 보행자 밀집
- 좁은 일방통행, 우회전 잦은 구간

## 6. 호송경비차량 요건 ★
- 운전자·금고실 분리
- 공격 견딜 구조
- 통신·경보·야간조명 장비

## 7. 호송경비 실시과정 핵심 ★
### (1) 경찰 협조 ★
- **출발 전일까지**
- **출발지·도착지 관할 경찰서장**
- **전자문서로 호송경비통지서 제출**

### (2) 인수·인계
- 신원 확인
- 수량·봉인 확인
- 서명·날인

### (3) 주의사항
- 적재·하역 신속
- 법정 경비용기 사용
- **열쇠 타인 보관 금지**
